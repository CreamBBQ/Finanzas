#### Inversión Bajo Incertidumbre 

COnsideramos una decisión de inversión donde I es el costo hundido, r la tasa de interes y $P_0$ al precio en el periodo cero (que es conocido por los inversores). Para el periodo siguiente el precio puede subir a $(1+u)P_0$ con probabilidad q o bajar en $(1-d)P_0$ con probabilidad 1-q. Luego, el precio se queda allí para siempre. Si definimos el valor presente esperado de la firma si invierte como $V_0$, entonces: 

$\begin{aligned} V_0 &=P_0+q \sum_{i=1}^{\infty} \frac{(1+u) P_o}{(1+r)^i}+(1-q) \sum_{i=1}^{\infty} \frac{(1-d) P_o}{(1+r)^i} \\ &=P_0\left\{1+\frac{1}{r}[q(1+u)+(1-q)(1-d)]\right\} \end{aligned}$

Tradicionalmente se solía comprara el valor presente con el costo de inversión y se concluía que si $v_0 > I$ se hacía la inversión y si no, no. Se notará desde ya que esto lleva a una pregunta fundamental en el mundo de las finanzas: ¿no tiene valor el esperar para invertir aún cuando el valor presente descontado ya es lo suficientemente grande para cubrir la inversión?

El inversor enfrenta un trade-off entre invertir hoy (y comenzar a cobrar el cash-flow) o esperar (considerando que puede subir el valor del proyecto) La decisión de esperar o no dependerá naturalmente de la apreciación esperada del proyecto y del costo de oportunidad de esperar. Notemos que si tenemos en cuenta el valor del proyecto mañana a precio $P_1$ esté será: 

$$ V_1(P_1) = P_1 + P_1 \dfrac{1}{(1+r)} + P_1 \dfrac{1}{(1+r)^2} + \cdots = P_1 \dfrac{1+r}{r} $$ 

Dado que a partir del período 1 todo permanece invariante, la decisión de inversión se tomará, ahora sí, solo si $V_1 > I$. En otras palabras, esperar ahora sí no tiene valor alguno: o se invierte ahora o no se invierte nunca. Luego, escribirmos el pago netro de la firma en el período 1 como: 

$F_1 = max[v_1 - I, 0]$ 

$F_1$ también se conoce como "valor de continuación en el período 0", pues es el valor que obtiene la firma si elige esperar. A su vez, el lado izquierda de esta ecuación es el valor de terminación en el periodo 1, pues es el valor que obtiene la firma si elige invertir en ese período (el valor de continuación aquí sería cero, porque sabemos que si no invierne en t=1 no invierte nunca). 

Ahora, sabiendo que en el período 1 va a comportarse óptimamente, ¿cuál será el problema de la firma en el período 0? Es importante notar que desde la perspectiva de t=0, el estado en el siguiente periodo es incierto. POr lo tanto el valor de la firma (es decir: el valor de la opción de invertir o esperar) debe ser: 

$$ F_0 = max\left\lbrace   V_0 - I, \dfrac{1}{(1+r)} E_0[F_1] \right\rbrace$$ 

Es decir: hoy la firma puede elegir entre invertir (lado izquierdo) o esperar, donde obtiene el valor descontado de la inversión futura. Naturalmente, $E_0[F_1]$ es: 

$$E_0[F_1] = q max\lbrace V_1 (P_0(1+u)-I, 0) \rbrace  + (1-q) max \lbrace V_1(P_0(1-d))- I, 0 \rbrace $$ 

Lo que se analiza cuando uno considera inversión bajo invectidumbre es una extensión a infinitos períodos del razonamiento que recién hicimos para dos períodos, donde en cada momento del tiempo nos preguntamos sin conviene esperar o invertir (y esto lo hacemos hasta que convenga invertir). En otras palabras: resolver ese problema es darle respuesta a la pregunta de cuándo es óptimo pagar un costo hundido I a cambio de obtener un proyecto cuyo valor es V y evoluciona estocásticamente. Algebraicamente la forma más simple de responder esa pregunta es utilizando un modelo que fue desarrollado por Siegel (1086). Suponemos que V evoluciona como una Brownian Motion geométrica definida como sigue: 

$$ d V = \alpha V dt + \sigma V dz $$ 

La posbilidad de estre proyecto cuyo valor evoluciona estocásticamente es lo que se conoce como una opción (en este caso, una especie de call). La firma tiene la opción (no la oblicación) de "comprar" el proyecto al precio I en cada período.  Veremos que la pregunta que queremos responde (en qué momento es óptimo invertir) es en un sentido análoga a responder cuál es el precio, en cada momento, de esta opción (¿cuánto debería valer, sin arbitraje, el derecho de comprar este proyecto si su valor actual es V?). 

 COnsideremos que el valor actual es V y que aún no se ha ejercido la opción. Entonces, el valor de la opción debe ser: 

 $$ F(V) = \max_T E[(V_T - I)e^{-\rho T} | V] $$ 

 Donde T es el momento en el que se realiza la inversión y $\rho$ es un factor de descuento (para que este problema tenga sentido debemos suponer que $\alpha < \rho $, de otra forma el proyecto nunca se realizaría pues se este se apreciaría más rápido que el costo de oportunidad y siempre convendría esperar).

##### Caso Determinístico 

Para presentar la mecánica del problema primer oresolvemos el problema determinístico, es decir, cuando la varianza es nula. Ahora la ecuación anterior se simplifica en: 

 $$ F(V_0) = \max_T E[(V_T(V_0) - I)e^{-\rho T}] $$ 

 Debemos encontrar el valor de $V_T$, lo cual implica resolver la ecuación diferencial $V'(t)= \alpha V(t)$ con condición inicial $V_0$. La solución de está ecuación es $V_t = V_0 e^{\alpha t} $. Por lo tanto el problema queda: 

 $$ F(V;T) = \max_T [(V e^{\alpha T} - I )e^{-\rho T}] $$ 

 De esta ecuación se desprende si $\alpha < 0 $ lo óptimo sería invertir inmediatamente si $ V > 0 $, o nunca realizar la inversión. Por lo tanto la pregunta de interés en esta sección ocurre en el caso en que $ \alpha \in (0, \rho) $. En dicho caso para obtener el máximo, derivamos con respecto a T: 

 $$ \dfrac{dF(V;T)}{dT} = ((\alpha - \rho)Ve^{\alpha T} + \rho I) e^{- \rho T} = 0 $$ 

 Conclutendo así que el tiempo óptimo es igual a: 

 $$ T* = \max \left \lbrace \dfrac{1}{\alpha}log\left(\dfrac{\rho I }{(\rho - \alpha)V} \right ) , 0  \right \rbrace $$

 Esta ecuación nos dice cuándo es óptimo invertir. También podemos ver al haber respondido a esta pregunta también hemos respondido a la pregunta del valor de la inversión. En particular, podemos ver que el tiempo óptimo es una función de la condición inicial V. Luego, nuestra función de valor inicial puede expresarse como: 

 $$ F(V) = (V e^{\alpha T*}) e^{-\rho T*} $$ 

 Que sólo epende del valor del proyecto V. Sabiendo esto podemos reveritr la pregunta: ¿Cuáles son los valors de V tales que deseamos invertir ahora? O, en otras palabras, cuáles son los valores de V tales que $T* = 0$? De la forma hallada para T*, podemos ver que el proyecto se ejecutraá inmediatamente siempre que el valor del proyecto sea mayor que un valor crítico $V*$: 

$F(V)=\left\{\begin{array}{c}{\left[\frac{\alpha I}{(\rho-\alpha)}\right]\left[\frac{(\rho-\alpha)}{\rho I} V\right]^{\rho / \alpha}} \\ V-I\end{array}\right.$
$\left.\begin{array}{r}\text { si } V \leq V^* \\ \text { si } V \geq V^*\end{array}\right\}$

 Este valor crítico nos permite respensar el probelam de la firma:e n lugar de plantearlo como "dentro de cuanto tiempo invertir dado que hoy el valor del proyecto es V", lo podemos pensar como un problema idéntico todo el tiempo (estacionario), en el cual en cada momento la firma se pregunta: ¿es suficientemente alto el valor del proyecto? Si lo es, decide invertir; si no lo es, decide esperar. Este razonamiento conlleva a que encontraremos un valor crítico $ V* > I $, lo cual nos permite ver lo que argumentamos al inicio de esta sección: invertimos cuando el valor presente del proyecto es estrictamente mayor al costo de invertir I, ya que se contempla el costo de oportunidad. Finalmente, reemplanzando T* en la ecuación de interés obtenemos el valor de la opción:

##### Caso Estocástico 

A continuación resolveremos el caso estocástico, es decir, cuando la varianza no es nula. Aquó no podremos derivar contra el tiempo óptimo de inversión como antes. Pero dadas las propiedades del problema sabes que este tiempo óptimo tendrá la forma "invertir siempre que el valor del proyecto pertenzca a una determinada región (en particular, que sea mayor a cierto valor crítico)".

Comenzamos entonces definidiendo a $\pi_t$ como el flujo de beneficios. Todavía no asumimos nada sobre las propiedades del poceso estocástico que genera a este flujo de beneficios. Luego, en todo momento, la firma enfrenmta una decisión de entrar o no entrar. Si entra, obtiene un valor de determincación $V(\pi)-I$, es decir, el valor presente de los flujos futuros dado que los beneficios corrientes son $\pi$, menos el costo de la inversión. Por otro lado, si no entra no obtiene cero (como lo dice el enfoque tradicional), sino que obtiene el valor de esperar. El mismo será respresnetado por el valor presente descontado de lvalor esperado de la opción el siguiente período. Entonces, dado que el valor de la opción $F(V)$ incorpora el hecho de que la firma toma la decisión óptima, dicho valor puede escribirse como: 

$$ F(V) = \max \left \lbrace V-I , \dfrac{1}{1 + \rho \Delta t}E[F(V + \Delta V)] \right \rbrace $$ 

(El lado izquierdo del operador de máximo nos muestra el pago por entrar hoy; el lado derecho el pago por esperar. Notemos que el rho es un factor de descuento que podría ser interpretado como una tasa de interés. El delta t nos dice que rango de tiempo estamos tomando, usualmente en los problemas de macroeconomía este rando es unitario (t = 1,2,3,..), así que ese delta t sería igual a uno y el lado derecho representaría el valor futuro de la opción en precios del tiempo presente)

La función F tendrá la siguiente forma: 


$F(V)= \left\{\begin{array}{c}{V-I} \\ ?\end{array}\right.$
$\left.\begin{array}{r}\text { si } V \leq V^* \\ \text { si } V \geq V^*\end{array}\right\}$

Tenemos ahora dos preguntas: que es F y que valor tiene V*. En la región incognita podemos ver la siguiente función: 

$$ F(V) = \dfrac{1}{1+\rho \Delta t} E[F(V + \Delta V)] $$ 

O como es equivalente: 

$$ (1+\rho \Delta t) F(V) = E[F(V + \Delta V)] $$ 

$$ F(V) + F(V) (\rho \Delta t) = E[F(V + \Delta V)] $$ 

$$ F(V) (\rho \Delta t) = E[\Delta F] $$ 

O su equivalente en tiempo continuo: 

$$ \rho F dt = E[dF] $$ 

Esta igualdad puede entenderse como una ecuación de equilibrio o no arbitraje, que se tiene que cumplir para que los inversores decidan adquirir un activo. El término de la izquierda es el retorno por unidad de teimpo que el inversor requiere para conservar el activo, es decir, el costo de oportunidad de mantener el proyecto en cartera. A la derecha tenemos la capitalización apreciación esperada del proyecto. 

Ahora, para contunidar, debemos asumir algo sobre el comportamiento de V, porque de otro modo no podríamos obtener soliciones. Asumamos entonces que sigue una Brownian Motion Geométrica. Aplicando el lema de Ito. F depende, en este caso, unicamente de V, por tanto: 

$$ dF = f_v (dv) + \dfrac{1}{2} f_{vv} (dv)^2 $$

Ahora bien, sabemos que: 

$$ dv = \alpha V dt + \sigma V dz $$

$$ \implies dF = f_v (\alpha V dt + \sigma V dz) + \dfrac{1}{2} f_{vv} (\alpha V dt + \sigma V dz)^2  $$ 

Desarrollando el término al cuadrado: 

$$ (dv)^2 = (\alpha V dt + \sigma V dz)^2 = (\alpha V dt)^2 + 2 \alpha V^2 dt dz \sigma + (\sigma V dz)^2 $$ 

Sabiendo que todo (dt)^m con m>1 tiende rapidamente a cero y que es despeciable para nuestro análisis concluímos que: 

$$ (dv)^2 = \sigma^2 V^2 dt $$ 

Luego, 

$$ dF = f_v (\alpha V dt + \sigma V dz) + \dfrac{1}{2} (\sigma^2 V^2 dt) $$ 

O lo que es equivalente: 

$$ dF = \left [\alpha F_v V + \dfrac{\sigma^2 F_{vv}V^2}{2} \right ]dt + F_v \sigma dz $$ 

Tomando esperanza matematica sabemos (y recordando que la misma es nula para dz): 

$$ E[df] = \left [\alpha F_v V + \dfrac{\sigma^2 F_{vv}V^2}{2} \right ]dt $$ 

Entonces, usando la condición de no arbitraje desarrollada anteriormente: 

$$ \rho F dt = \left [\alpha F_v V + \dfrac{\sigma^2 F_{vv}V^2}{2} \right ]dt $$ 

$$ \implies \rho F  = \left [\alpha F_v V + \dfrac{\sigma^2 F_{vv}V^2}{2} \right] $$

Reordenando los términos podemos expresar todo en una ecuación diferencial de segundo orden homogenea: 

$$ \alpha V F_v + \dfrac{\sigma^2 F_{vv}V^2}{2} - \rho F = 0 $$ 

Cuya solución general, usando conocimientos de ecuaciones diferenciales es: 

$$ F(V) = A_1 V^{\beta_1} + A_2 V^{\beta_2} $$

donde $\beta_1 > 1$ y $\beta_2 < 0$. 

Resumiendo, nuestro problema ahora se redujo a: 

$F(V)= \left\{\begin{array}{c}{V-I} \\ {A_1 V^{\beta_1} + A_2 V^{\beta_2} }\end{array}\right.$
$\left.\begin{array}{r}\text { si } V \leq V^* \\ \text { si } V < V^*\end{array}\right\}$


Ahora bien, como sabemos esto no es una solución cerrada pues quedan $A_i$ y V* por determinar. Tenemos que impartir tres condiciones terminales para hallar una solución al problema. Estas son presentadas a continuación: 

<u> Opción bajo valor nulo </u>

Cuando el valor del proyecto es nulo la opción por esperar también debe serlo: 

$F(0) = 0$ 

De modo que, si tomamos límite a la parte inferior de la función partida sabemos que $A_2 = 0$, porque de lo contrario el término $V^{\beta_2}$ tendería a infinito y la condición no se cumpliría (recordemos que $\beta_2$ es negativo). 

<u>Value matching </u>

Debemos pedir continuidad en la función F, porque de lo contrario habría inconsistencia temporal en la decisión del individuo. Esto nos lleva a afirmar que debe ser cierto: 

$ V^* - I = A_1 V^{*\beta_1} $ 

<U> Smooth pathing </u>

Al momento de inveritr una suba en V le agrega marginalmente tanto al valor del proyecto como a la opción de esperar, entonces debe ser cierto que: 

$ 1 = A_1 \beta_1 (V^*)^{\beta_1 -1} $ 

Notar que esto es equivalente a plantear que la derivbada de ambas ramas deben ser iguales 

<u> Solución cerrada </u>

Tenemos entonces un sistema de dos ecuaciones con dos incognitas que arrojan los siguientes resultados: 

$V^* = \dfrac{\beta_1}{\beta_1 - 1 } I $ 

$A = \dfrac{V^* - I }{(V^*)^{\beta_1}} $

Podemos ver entonces que tanto la incertidumbre como la irreversibilidad de la decisión de invertir abren una brecha entre el valor óptimo de invertir y su costo, demostrando que la visión clásica es imcompleta y propone una regla de inversión que no es correcta. 







#### Salida óptima

En la sección anterior estudiamos el caso de un inversor que tenía que decidir el momento óptimo de invertir en un proyecto dado, para posteriormente quedarse en ese proyecto para siempre. Supongamos ahora que, una vez que estamos adentro, tenemos la opción de salir. Para ver como se modifica el problema, pensemos en el problema que tendría la firma una vez que esté operativa: la misma enfretaria la decisión de salir o quedarse. Para estudiar este problema, supongamos que los beneficios siguen un proceso browniano: 

$ d \pi = \pi \mu dt + \pi \sigma dz $

Para evaluar el proyecto necesitamos una condición de no arbitraje entre los mercados. Esta será: 

$ \rho V dt = \pi + E_t[dV] $ 

-------------------------------------------------

<u> Disclaimer: desarrollo de la condición de no arbitraje </u>

Consideremos el valor del proyecto calculado como el valor esperado descontado de las ganancias futuras. Sea $\pi$ el flujo de beneficios, entonces el valor del proyecto será: 

$ V_t = E_t \left [ \sum_{i=0}^n (\dfrac{1}{1 + \rho}^i) \pi_{t+i} \right]  $

Que puede ser reescrita como: 

$ V_t = \pi + \dfrac{1}{1+\rho} E_t [V_{t+1}] $

Ahora queremos llevar esta ecuación a tiempo continuo. Para hacer esto debemos llevar el intervalo de tiempo de 1 a dt: 

$ V = \pi dt + \dfrac{1}{1+\rho dt} E_t [V_{t+dt}] $

$ \implies V(1 + \rho dt) = \pi dt (1 + \rho dt) + E_t[V_{t + dt}] $ 

$ \implies \rho V dt = \pi dt + E_t[dV] $

Notemos que esta ecuación solo no está diciendo que podría invertir V unidades de dinero a una tasa $\rho$ en el mercado de bonos y eso debería ser igual a estar en el proyecto o si no el mercado no estría en equilibrio. En el fodo también nos está diciendo que el mercado y las personas descuentan a la misma tasa. 

-------------------------------------------------

Como sabemos que el valor del proyecto es una función de los beneficios y dado que los beneficios siguen una brownian motion geometrica podemos obtener el cambio en el valor del proyecto simplemente aplicando el Lema de Ito: 

$ dV = V_\pi d \pi + \dfrac{V_{\pi \pi} (d \pi)^2}{2} $

Sabiendo el movimiento de $\pi$ podemos reescribir la ecuación anterior como: 

$ dV = V_\pi  (\pi \mu dt + \pi \sigma dz )+ \dfrac{V_{\pi \pi} (\pi \mu dt + \pi \sigma dz )^2}{2} $

Recordando que existen terminos que se van muy rapodamente a cero (todo $dt^n$ con $ n>1 $) y que $(dz)^2 = dt$, obtenemos: 

$ dV = V_\pi (\pi \mu dt + \pi \sigma dz) + \dfrac{V_{\pi \pi} (\sigma \pi)^2 dt}{2} $

Tomando esperanza matemática: 

$ E[dV] =  \left [V_\pi (\pi \mu) + \dfrac{V_{\pi \pi} (\sigma \pi)^2}{2} \right ] dt$

Usando la condición de no arbitraje y reemplazando la esperanza antes conseguida: 

$  \rho V dt = \pi dt + \left [V_\pi (\pi \mu) + \dfrac{V_{\pi \pi} (\sigma \pi)^2}{2} \right ] dt $

Luego obtenemos una ecuación diferencial de 2do grado no homogénea para el valor del proyecto: 

$ \pi = \rho V - \left [V_\pi (\pi \mu) + \dfrac{V_{\pi \pi} (\sigma \pi)^2}{2} \right ] $

La solución general de dicha ecuación la podemos escribir como la suma de la solución general de la homogenea sumada a la solución particular de la no homogenea. No se va ahondar en las distintas implicancias de las ecuaciones, sino que solo se van a proponer "guess" para cada parte. 

<u> Parte homogenea </u>

$ pV - V_\pi \mu \pi - \dfrac{\sigma^2 \pi^2 V_{\pi \pi}}{2} = 0 $

Guess: 

$ V^{SGH} = B \pi^\beta $

$ V^{SGH} = B_1 \pi^\beta_1 + B_2 \pi^\beta_2$

Con $B_1$ y $B_2$ a determinar y $\beta_1 >1$ y $\beta_2 < 0$

<u> Parte no homogenea </u>


$ \rho V - V_\pi \mu \pi - \dfrac{\sigma^2 \pi^2 V_{\pi \pi}}{2} = \pi $

Guess: 

$ V^{SPNH} = k \pi  $

De modo que: 

$ \rho(k \pi) - k \mu \pi - \dfrac{\sigma^2 \pi^2 (0)}{2} = \pi $

$ \rho(k) - k \mu  = 1 $

$\implies k = \dfrac{1}{\rho - \mu} $

$ \implies V^{SPNH} = \dfrac{1}{\rho - \mu}\pi $

<u> Solución general </u>

Como anticipamos, la solución general de la ecuación diferencial va a estar dada por: 

$ V = \dfrac{\pi}{\rho - \mu} +  B_1 \pi^\beta_1 + B_2 \pi^\beta_2 $

- La parte $\dfrac{\pi}{\rho - \mu}$ es lo que se como valor fundamental del proyecto 

- Mientras que $ B_1 \pi^\beta_1 + B_2 \pi^\beta_2 $ computa el valor de la opción de salida (notar que la forma funcional de la opción de netrar o salir son idénticas (no así los coeficientes))

Si bien se tiene ahora la forma funcional del valor del proyecto, lo cierto es que aún quedan incognitas por resolver. En la siguiente sección se desarrollan las mismas.






##### Condiciones de borde: caso entrada y salida

Cuando se consideró el caso en que no se podía salir del proyecto se llegó a una solución para la opción de entrar al proyecto que dependía del valor (V) del mismo. Para actualizar esta función al contexto en el que existe la opción de salir debemos expresarla en terminos de los beneficios ($\pi$) de la siguiente forma: 

$ \rho F^{ent}(\pi) = E[dF^{ent}(\pi)] $ 

De esta ecuación de no arbitraje se tienen que notar dos cosas: 

- Que en la sección donde no había opción de salida también se pudo haber llegado a la solución utilizando una función del valor de entrar al proyecto dependiendo de los beneficios y no solo es valido para la función que debende del valor del proyecto 
- Que $d \pi = \pi \mu dt + \pi \sigma dz $, por lo que se puede usar lema de Ito para obtener la ecuación diferencial que resuelve la función. De acá se sigue que la función F deberá tener la forma con un proceso igual (pero con parámetros distintos)

Con eso en mente, sabemos que la ecuación diferencial que resuelve $F^{ent}$ viene dada por: 

$ \dfrac{\sigma^2 F^{ent} \pi^2}{2} + \mu F^{ent} \pi - \rho F^{ent} = 0 $ 

Por la misma razón que antes, la solución tiene forma general: 

$ F^{ent}(\pi) =  \tilde{A_1} \pi^{\beta_1} + \tilde{A_2 }\pi^{\beta_2} $ 

Habiendo desarrollado las formas funcionales de las opciones de entrada y salida nos resta encontrar seis parámetros (a saber: $ \tilde{A_1}, \space \tilde{A_2},\space  B_1, \space B_2, \space \underline{\pi}, \space  \overline{\pi}$. donde las últimas dos son el valor crítico de salida y entrada del proyecto respectivamente). A continuación se dan las seis condiciones terminales que dan solución cerrada al problema. 

<u> Valor de entrada nulo tienden a cero si los beneficios Tienden a cero </u> 

si $ \pi \rightarrow 0 $

$ \implies F^{ent} \rightarrow 0 $ 

$ \implies \tilde{A_2} = 0 $ 

$ \tilde{A_2} $ debe ser igual a cero para que la opción de entrar no tenga valor si los beneficios teinden a cero. 

<u> Valor del proyecto infinito si los beneficios tienden a infinito </u> 

si $ \pi \rightarrow \infty $ 

$ \implies V (\pi) \rightarrow \infty $ 

$ \implies F^{sal} = 0 $ 

$ \implies B_1 = 0 $ 

(nadie quiere salir de un proyecto que paga infinito) 

<u> Value matching </u>

Sea $\underline{\pi}$ el valor de los beneficios para los cuales es óptimo salir del proyecto y $\overline{\pi}$ donde es óptimo entrar, planeamos las siguientes condiciones: 

- (entrada) $ V(\overline{\pi}) = F^{ent}(\overline{\pi}) + I $; El valor del proyecto evaluado en el valor crítico de entrada es igual a la opción de entrar en ese momento más el costo hundido 
- (salida) $ V(\underline{\pi}) = F^{ent}(\underline{\pi}) + E $; El valor del proyecto evaluado en el valor crítico de salida es igual a la opción de entrar en el proyecto en ese momento más E, que representa lo que me pagan por vender el proyecto

<u> Smooth pathing </u>

Marginalmente los cambios en los distintos pagos deben ser iguales

-(entrada) $ V_\pi (\overline{\pi}) = F^{ent}_\pi (\overline{\pi}) $

-(salida) $ V_\pi (\underline{\pi}) = F^{ent}_\pi (\underline{\pi}) $

<u> Solución cerrada </u>

Con estas condiciones se encuentra una solución cerrada al sistema de ecuaciones planteado. Esta solución generalmente no es numérica, sino que tiene que ser resuelta computacionalmente.
