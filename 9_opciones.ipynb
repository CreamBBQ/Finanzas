<h1> <center> OPCIONES </H1> </center>

En la corriente sección vamos a analizar instrumentos financieros que comúnmente se denominan como derivados. Estos instrumentos consisten por lo general en contratos cuyo precio depende de las características salientes de uno o más activos. Una vez definidos el tipo de contrato, la tarea mayor es encontrar el precio justo de dicho contrato. Las estrategias para encontrar dicho precio por lo general se basan en que las propiedades estocásticas del precio del contrato se derivan de las del activos principal (lema de ITO). 


Notación: 
 - Underlyng asset (activo subyacente): activo sobre el que se escribió la opción. 
 - $S_\tau$ es el precio del subyacente en el momento $\tau$ 
 - Expiration day (T): día en el que se puede ejectuar la opción 
 - Strike price (X): precio al que el comprador estima que se va a ir el activo subyacente  

<u> Call </u>

Una opción Call es el derecho (pero no la obligación) a comprar un activo particular a un precio previamente determinado en un momento futuro especificado 

Llegado el momento T: 
- Si $S_T  > X$ se ejecuta el call
- si $S_T  < X$ no se ejecuta el call 

El poseedor del call está interesado en que el subyacente suba de precio. De esta forma pagaría $X$ por el activo y lo podría vender al precio de mercado $S_T$ obteniendo así un margen de ganancia. De modo que llegados al momento T el payoff del call es: 

$$ C_T = máx \{S_T - X, 0 \} $$ 

![](https://i.imgur.com/yczpbAn.png)

<u> Put </u>

Una opción Put es el derecho (no la obligación) a vender un activo a un precio determinado previamente, en un momento futuro especificado. 

Llegado el momento T: 
 - Si $S_T > X$ no se ejecuta el put 
 - Si $S_T < X$ se ejecuta el put 

Luego, a madurez T el payoff del put es: 

$$ p_T = más \{ X - S_T, 0 \} $$ 

El poseedor del put está interesando en que el el subyacente baje de precio, de este modo puede vender el producto a un precio más alto que el de mercado, obteniendo así una ganancias. 

![](https://i.imgur.com/27wGmIZ.png)

##### El precio justo 

La pregunta más importante a responder es: ¿Cuál es el precio justo al que se debe tranzar el contrato a tiempo t? Para determinar dicho precio notamos que la opción, por ejemplo un Call, depende de: 

1. El valor del stock a tiempo t, $S_t$, y de la posible evolución del mismo desde t hasta el día de expiración. 
2. Cuánto falta para el día de expiración, T-t
3. El strike price 
4. La volatilidad del stock price 
5. La tasa libre de riesgo 
6. Los dividendos esperados que pagará el stock durante la vida de la opción 

Por lo tanto, si se denota C(.) al precio del call, este depende de los siguientes argumentos: $C(S_t, \space X, \space T-t, \space \sigma_S, \space r)$. Dado que el precio del subyacente se mueve de forma estocástica, es lógico pensar que el precio del Call también dependa del riesgo del precio del Stock y de su tendencia, es por esto que el precio del Call también es una función de: $\sigma_s$ y r. 

Notar que: 

1. En lugar de aparecer la tendencia del activo aparece la tasa libre de riesgo (el porqué se discutirá más adelante)
2. Como el call está acotado por debajo uno esperaría que cuando sube la volatilidad del subyacente, suba el precio del call (si el precio del activo cae demasiado, el comprador del call no pierde más que lo que pagó por el derivado, mientras que si sube mucho podría ganar mucho más, es por esto que decimos que el precio debe subir)
3. A tiempo t=T el precio del call debe de ser igual al pay-off del Call pues de otra manera habría oportunidad de arbitraje. 

Veremos a continuación que la determinación de $C(S_t, \space X, \space T-t, \space \sigma_S, \space r)$ es el desafio mayor al que nos enfrentamos cuando estudiamos opciones. CUando no hacemos supuestos distributivos sobre el movimiento del precio del subyacente solo podemos encontrar cotas superiores e inferiores de los valores del Call y del Put. Por otro lado, bajo ciertos supuestos sobre la evolución de dicho precio, podemos encontrar formas reducidas exactas para el precio de las opciones (usaremos lema de Ito con Brownian Motion geométrica o aritmetica). 

##### Cotas: Call de un subyacente que no paga dividendos 

<u> Cota superior </u>

El precio de un call es siempre menor al del stock, en caso contrario todos venderían el call y comprarían el stock. 

$$ c(.) \leq S $$

Ejemplo: supongamos por un momento que la afirmación anterior no es cierta. Se tiene entonces que al día de hoy la acción de Apple vale 4 y yo escribo un call con strike price 5, a precio 8 y día de expiración mañana. Juan me lo compra. Recibo 8 y compro 2 acciones de Apple el día de hoy. Mañana pudieron haber pasado dos cosas: 

- La acción subió a 6. Mis acciones de Apple ahora valen 12. Debo vender una de ellas a Juan a precio 5. Tengo entonces 12 (porque aún poseo una acción) y 5 que me paga Juan por la acción. He ganado 17. 

- La acción bajó a 4. Mis acciones de Apple ahora valen 8, como Juan no ejecutará su call esto quedará para mí. 

De esta forma, vender un Call (ir short en el y comprar el activo) siempre dará un pago positivo por lo que el mercado no permitiría este arbitraje. 


<u> Cota inferior </u>

Un call como los vistos hasta ahora (que no paga dividendos) verifica: 

$$ c(.) \geq S - X e^{-r(T-t)} $$

Para ver esto, consideremos los siguientes portafolios: 

1. $ \Pi_1 = c + X e^{-r(T-t)} $ ; va long en el call e invierte la cantidad de dinero justa en el activo libre de riesgo para que en el momento T le pague X
2. $ \Pi_2 = S $ ; compra el stock

Al momento T el pago del portafolio 1 es:  

- Si $S_T \geq X $ el pago será: $ (S_T -X)+X = S_T $
- Si $S_T \leq X $ el pago será $ 0 + X = X $ 

O lo que es equivalente: $ máx \{S_T, X \}$. 

Mientras que el portafolio 2 tiene como pago $S_T$, el valor del stock en el momento T. Se verifica entonces: 

$$ \Pi_1(., T) \geq \Pi_2(., T) $$ 

El portafolio 1 siempre tiene más ganancia que el 2. Para evitar posibilidades de arbitraje, el portafolio 1 siempre debe ser más caro que el portafolio 2 en todo momento del tiempo: 

$$ \Pi_1 \geq \Pi_2 $$

$$ c + X e^{-r(T-t)} \geq S $$ 

$$ c \geq S - X e^{-r(T-t)} $$




##### Cotas: Put de un subyacente que no paga dividendos 

<u> Cota superior </u>

Teniendo en cuenta que el valor del put a tiempo T es $máx \{X-S, 0 \}$, debe ser cierto que $ p(.) \leq X e^{-r(T-t)} $.

Imaginemos que esto no es cierto así que escribo un put a un precio mayor a $X e^{-r(T-t)} $, invierto eso en el activo libre de riesgo de modo que en el momento T tengo más dinero que X. Si el put se ejecuta tengo de ganancia la diferencia entre el valor en el que vendí el put y X, si no se ejecuta tengo de ganancia todo lo que invertí en el activo libre de riesgo. Este arbitraje no es permitido por el accionar de los mercados competitivos. 

<u> Cota inferior </u>

Un put que no paga dividendo verifica: 

$$ p(.) \geq X e^{-r(T-t)} - S $$ 

Siguiendo un argumento similar al del call, consideremos los siguientes portafolios: 

$$ \Pi_1 = p + S $$

$$ \Pi_2 = X e^{-r(T-t)} $$ 

Para t = T el valor del portafolio 1 puede ser: 
- Si $S_T < X$ el portafolio vale $(X-S_T) + S_T = X$
- Si $S_T > X$ el portafolio vale $S_T$ 

O lo que es equivalente: $\Pi_1(.,T) = máx \{X, S_T\}$

Mientras que en t = T el valor del portafolio 2 es $\Pi_2(.,T) = X$

Entonces es cierto que: 

$$\Pi_1(., T) \geq \Pi_2(., T) $$

Luego, para que no exista posibilidad de arbitrar, en todo t se verifica que: 

 $$\Pi_1 \geq \Pi_2$$
 $$ p + S \geq Xe^{-r(T-t)} $$
 $$ p \geq Xe^{-r(T-t)} - S $$


##### Cotas: Call de un subyacente que paga dividendos 

Examinemos cómo se modifican estos resultados si el stock subyacente paga dividendos. Llamemos D al valor presente de los dividendos durante la vida de la opción. 

<u> Cota inferior </u>

Un call de estas características verifica: 

$$ c(.)  \geq S - D - Xe^{-r(T-t)} $$ 

Para ver esto consideremos los siguientes portafolios: 

$$ \Pi_1 = c + D + Xe^{-r(T-t)} $$

$$ \Pi_2 = S $$ 

Para t = T el valor de los portafolios son: 

$$ \Pi_1(., T) = máx \{S_T, X \} + D$$

$$ \Pi_2(., T) = S_T + D $$

Dado que $\Pi_1(., T) \geq \Pi_2(., T)$ por lo arbitraje se verifica que para todo t: 

$$ c + D + Xe^{-r(T-t)} \geq S $$ 


##### Cotas: Put de un subyacente que paga dividendos 

Un put de estas características verifica que: 

$$ p(.) \geq X e^{-r(T-t)} + D - S $$ 

Para ver esto consideremos los siguientes portafolios: 

$$ \Pi_1 = p + S $$

$$ \Pi_2 = D + X e^{-r(T-t)} $$

Para t = T el valor de los portafolios es: 

$$ \Pi_1(., T) = máx \{S_T, X \} + D $$ 

$$ \Pi_2(., T) = X + D $$ 

Por lo que es cierto que: $ \Pi_1 \geq \Pi_2(., T) $ 

Entonces, para que no exista posibilidad de arbitrar se verifica que para todo t: 

$$ p + S \geq D + X e^{-r(T-t)} $$ 

$$ p \geq D + X e^{-r(T-t)} - S $$

##### Opciones Americanas 

Las opciones vistas hasta ahora se conocen como *opciones europeas* y tienen la característica solo pueden ejercitarse en la fecha de expiración. Las *opciones americanas* pueden ejecutarse en cualquier momento antes de la fecha de expiración. Existen ciertas propiedades que relacionan (o no) las opciones europeas de las americanas. Las examinamos a continuación: 

Proposición: nunca se ejecuta un call americano que no paga dividendos antes del tiempo en que la opción expira (de esta forma un call europeo y americano son en esencia lo mismo)

demostración: consideremos los siguientes portafolios: 

$$ \Pi_1 = C + X e^{-r(T-t)} $$ 

$$ \Pi_2 = S $$ 

Si el call se ejecuta en $ \tau  < T $ los payoffs de los portafolios son: 

$$ \Pi_1(., \tau) = S - X + X e^{-r(T-\tau)} $$ 

$$ \Pi_2(., \tau) = S $$ 

Por lo que $ \Pi_2(., \tau) \geq \Pi_1(., \tau) $


Si se ejecuta en T los payoffs son: 

$$ \Pi_1(., T) =máx \{S_T, X \} $$ 

$$ \Pi_2(., \tau) = S_T $$ 

Por lo que $ \Pi_1(.,T) \geq \Pi_2(.,T) $

Concluimos entonces que nunca es óptima ejecutar este tipo de calls antes de su fecha de expiración. Luego, el precio de un call americano y de un call europeo sobre un stock que no paga dividendos debería ser el mismo. 

Proposición: puede ser óptimo ejecutar un put antes del tiempo en que la opción expira

Demostración: Consideremos los siguientes portafolios

$$ \Pi_1 = p + S $$ 

$$ \Pi_2 = Xe^{-r(T-t)} $$ 

Si el put se ejecuta en $ \tau < T $ los payofss de los portafolios son: 

$$ \Pi_1(., \tau) = (X-S)+S = X $$ 

$$ \Pi_2(., \tau) = Xe^{-r(T-\tau)} $$ 

Por lo que $ \Pi_1(., \tau) > \Pi_2(., \tau) $

Si se ejecuta en T los payoffs son: 

$$ \Pi_1(., T) = máx \{S_T, X \} $$ 

$$ \Pi_2(., T) = X $$ 

Por lo que $ \Pi_1(., T) \geq \Pi_2(., T) $

No podemos excluir la posibilidad de ejecutar la opción antes de su vencimiento. Es esperable que el precio de un put americano sea mayor al precio de un put europeo (dado que el americano ofrece más posibilidades al inversor). 

#### Paridad Call-Put (europeos)

Relación entre el precio de un call y un put con misma madurez y strike price. Consideremos los siguientes portafolios: 

$$ \Pi_1 = c + Xe^{-r(T-t)} $$

$$ \Pi_2 = p + S $$ 

En la fecha de expiración el payoff de ambos portafolios es el mismo: 

$$ \Pi_1(., T) = \Pi_2(., T) = máx \{S_T, X \} $$

Entonces a tiempos T debe ser cierto: 

$$ c + X = p + S $$ 

Para que no exista posibilidad de arbitrar, esta igualdad se cumple para todo momento t: 

$$ c + X e^{-r(T-t)} = p + S $$

La igualdad anterior recibe el nombre de "paridad call-put" y supone una relación de no arbitraje entre el precio de ambas opciones. 

Notar que de acá también se puede scar una relación entre call y put americanos: 

$$ c_e + X e^{-r(T-t)} - S = c_a + X e^{-r(T-t)} - S $$

Por el resultado de igualdad de precio entre un call americano y europeo que derivamos anteriormente. No obstante, también demostramos que: $p_a > p_e $ porque daba más posibilidades al inversor. De esta forma debe ser cierto que: 

$$ p_a  >  c_a + X e^{-r(T-t)} - S $$ 





#### Derivación del precio de una opción 

Analicemos cómo encontrar el precio de un call, c. Asumimos que el precio del stock sigue una distribución binomial. Así, el precio del Stock, S, puede: 

- Subir a un nivel uS, con $u > 1$, con probabilidad q
- Bajar a un nivel dS, con $d < 1$ con probabilidad 1-q

Asumimos que $ d < r < u $. r es igual a 1 más la tasa de interés. De esta forma ni el riesgoso ni el libre de riesgo domina estocasticamente de primer orden al otro. Para determinar el precio consideramos dos enfoques distintos, pero equivalentes: 
1. Replicar el pago del call con un portafolio.
2. Valor presente descontado bajo neutralidad al riesgo. 

<u> Enfoque I: Replicar el payoff </u>

En el período siguiente el precio de la opción depende de la realización de S y será: 

$$ c_u = más \{uS - X, 0 \} $$

$$ c_d = más \{dS - X, 0 \} $$

La pregunta clave es: ¿es posible replicar el payoff del call en cambios estados tranzando el stock y el libre de riesgo? Construimos un portafolio de una cantidad $\Delta$ del stock y Z puesta a la tasa libre de riesgo. Queremos encontrar $\Delta$ y Z para reproducir el payoff del call em ambos estados de la naturaleza, es decir: 

$$ \Delta S + Z r = c $$ 

El sistema que queremos resolver es entonces

$$ \Delta uS + rZ = c_u $$ 

$$ \Delta ds + rz = c_d $$ 

La solución a este sistema da las siguientes cantidades que reproducen el payoff del call: 

$$ \begin{bmatrix} \Delta \\ Z  \end{bmatrix} = \dfrac{1}{(u-d)rS} \begin{bmatrix} r & -r \\ -dS & uS \end{bmatrix} \begin{bmatrix} c_u \\ c_d  \end{bmatrix}$$

Sustituyendo en el portafolio obtenemos: 

$$ c = \dfrac{1}{r} \left (\dfrac{r-d}{u-d} c_u + \dfrac{u-r}{u-d}c_d \right) $$

Es decir, bajo este modelo el precio del call es un promedio ponderado descontado del payoff del call en cada estado. 

<u> Enfoque II: Valuación bajo neutralidad al riesgo </u>

Vamos a realizar la evaluación del precio de la opción bajo el supuesto de que los agentes son neutrales al riesgo. Vamos a encontrar que bajo dicha medida de probabilidad, el precio que obtenemos es el mismo que aquel que obtuvimos cuando valuamos la opción reproduciendo todos los estados de la naturaleza- De lo anterior se desprende que si nosotros podemos obtener, de alguna forma, probabilidades de los estados de la naturaleza cuando (como si) los agentes son (fueran) neutrales al riesgo (aunque no lo sean), siempre vamos a poder utilizar este método de valuación independientemente de la aversión al riesgo de los mismos. 

Supongamos ahora que los agentes son neutrales al resigo. Sabemos que bajo dicho escenario, que todos los activos deben crecer a la misma tasa promedio, e igual a la tasa libre de riesgo, entonces 

$$ rS = puS + (1-p) dS $$ 

Donde p es la probabilidad, bajo el supuesto de que los agentes son neutrales al riesgo que en general es distinta de q (que el la probabilidad de que el stock suba o baje). Entonces podemos encontrar el valor de p que satisface que los agentes son neutrales al riesgo (o que todos los activos crecen a la misma tasa), 

$$ p = \dfrac{r-d}{u-d} $$

$$ (1-p) = \dfrac{u-r}{u-d} $$ 

Ahora si los agentes son neutrales al riesgo, podemos calcular el valor de la opción a tiempo t como el valor descontado (esperado) del payoff, es decir: 

$$ c=\dfrac{1}{r}(pc_u + (1-p)c_d) = \dfrac{1}{r} \left [\dfrac{r-d}{u-d}c_u  + \dfrac{u-r}{u-d}c_d  \right] $$

La conclusión más importante que podemos obtener a esta altura es que el precio que obtenemos es exactamente el mismo que obtenemos utilizando argumentos de arbitraje, que es valido independientemente de las preferencias hacia el riesgo de los individuos. Por lo tanto siempre se puede encontrar el precio de un derivado como si los individuos fueran neutrales al riesgo, para ello siempre es necesario encontrar las probabilidades bajo neutralidad al riesgo. 


#### Modelos binomial de varios periodos 

La forma de encontrar el precio de una opción cuando consideramos más de un periodo es análoga. Uno puede hacer nuevamente el portafolio replicado o puede descontarlo "como si los agentes fueran neutrales al riesgo" y ambos resultados serán iguales. Consideremos el siguiente árbol para la evolución de S: 

![](Captura%20de%20pantalla%20de%202022-11-08%2009-24-19.png)

Si en la primera realización el estado es uS los posibles payoffs en el segundo periodo son: 

$$ c_{uu} = más \{u(uS)-X, 0 \} $$ 

$$ c_{ud} = máx \{d(uS)-X, 0 \} $$

Por otro lado si la primera realización es dS los posibles payoffs en el segundo periodo son: 

$$ c_{du} = más \{u(dS)-X, 0 \} $$ 

$$ c_{dd} = máx \{d(dS)-X, 0 \} $$

Por lo tanto usando los argumento comunes para replicar un portafolio (o el descuento usando la probabilidad neutral al riesgo) obtenemos: 

$$ c_u = \dfrac{1}{r} [pc_{uu} + (1-p)c_{ud}] $$

$$ c_d = \dfrac{1}{r} [pc_{ud} + (1-p)c_{dd}] $$ 

Ahora repitiendo el argumento de arbitraje (o descontando otra vez con las probabilidades neutrales al riesgo) obtenemos el precio en tiempo 0: 

$$ c = \dfrac{1}{r^2} [p^2 c_{uu} + 2p(1-p)c_{ud}+(1-p)^2c_{dd}] $$

Generalizando a n periodos obtenemos: 

$$ c = \dfrac{1}{r^n} \sum_{i=0}^n \dfrac{n!}{(n-i)!i!} p^i (1-p)^{n-i}c(u^i, d^{n-i}) $$

Donde $ c(u^i, d^{n-i}) = máx \{S u^i d^{n-i} - X, 0 \} $
