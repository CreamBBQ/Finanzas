<h1> <center> BONOS CON CUPONES </H1> </center>

Lo cupones generalmente se especifican como fracciones el monto principal. Se puede pensar al bono con cupones como un portafolio de zero-coupon bonds, donde cada cupón puede ser interpretado como el cobro de un cono con cero cupÓn (CC) en el intervalo que corresponde al cupón que se paga en ese periodo, es decir, el monto principal del bono CC es igual al cupón. Asi formamos el flujo de pago de cupones con n bonos CC. Finalmente tomamos un bono CC final con la misma madurez que el bono inicial, que represente el pago del principal. Otra forma de pensarlo es suponer que los cupones se pagan al final del periodo todos juntos, llevados a valor futuro se suman al monto principal y se hace el pricing del bono como si no pagara cupones (pero con un monto principal mayor) En otras palabras, siempre existe un cero coupon bond que sea equivalente (en precio y retorno) al bono que paga cupones. 

Supongamos que el valor del bono B es igual a la suma de los cupones que se pagan en cada periodo, descontados usando la tasa interna de retorno del bono: 

$$ B = \sum_{i=1}^n C_i exp\{-yt_i\} $$ 

Cuando estamos buscando el precio de un bono en modelos de derivados como los vistos anteriormente, simplemente suponemos que los intervalos de tiempo son infinitesimales: dt. y en cada uno de estos intervalos se recibe un cupón de monto K(r,t). Los cupones entran en la ecuación del precio de un bono de forma similar a como entraban los dividendos en la ecuación del precio de una opción: 

$$ P_t + \dfrac{1}{2}w^2 P_{rr} + (u - \lambda w) P_r + K(r,t) - rP = 0 $$ 

Más allá del precio del bono, ya derivado, el poseedor de una cartera de bonos que pagan cupones generalmente está interesado en protegerse ante cambios grandes en el valor de los mismos. 

Para ver como cubrirnos el riesgo de la tasa de interés, lo primero que tenemos que hacer es saber como cambios en la misma afectan el valor de los distintos bonos. Comenzaremos analizando el problema bajo el supuesto de que la curva de rendimiento es flat (esto quiere decir que el rendimiento es igual para cualquier madurez, es decir, todos tendrían la misma tasa de retorno anual) y que toda la información se obtiene usando la tasa interna de retorno del bono o la YTM, yield to maturity que será descrita a continuación 

#### Tasa interna de retorno (TIR)

La tasa interna de retorno ("yield to maturity", YTM) puede pensarse como la tasa de interés consistente con el precio actual asumiendo que es la única tasa para todo el horizonte. 

$$ P_t = P(y) = \sum_{i=1}^m \dfrac{C_i}{(1+y)^{T_i - t}} $$

Básicamente, la YTM es una promesa de rendimiento, porque supone que el bono será mantenido hasta su madurez y que todos los cupones podrán ser re-invertidos a la YTM. Si el inversor decide gastar esos cupones en otra cosa o los re-invierte a una tasa distinta a la YTM, el rendimiento final de dicho bono diferirá de la YTM. Dado que en general no todos los cupones pueden ser re-invertidos a la YTM, el rendimiento final del bono no será igual al calculado ex-ante. En general, los bonos de mayor madurez y aquellos que con mayores cupones conllevan un mayor riesgo de inversión. A medida que cae la tasa del mercado, el precio de los bonos sube, y vice versa. 

-------------------------------------

Como podemos imaginar el supuesto de que toda la información está contenida en la YTM es bastante fuere. Lo que pasa en la "realidad" es que el precio del bono está dado por: 

$$ P_t = \sum_{i=1}^m C_i B(t,T_i) = \sum_{i=1}^m \dfrac{C_i}{[1 + R(t,T_i-t)]^{T_i - t}} $$ 

Donde R es la tasa a la que se descuentan los distintos B bonos sin cupones. COmo podemos ver el precio es una función de m distintas tasas de interés (variables) y de la madurez del portafolio. Esto hace que el precio del bono dependa de un gran número de factores de riesgos, por lo que en la práctica es difícil (o costoso) eliminar el riesgo de todos estos factores. En la práctica se intenta simplificar el manejo de los portafolios reduciendo el número de variables riesgosas. 

-------------------------------------------


Por ejemplo, consideremos un bono de dos años con un valor nominal de $100 que paga un cupón de 2% anual que se paga semestralmente. Este cupón se vende en el mercado por 90.9. La yield se obtiene de: 

$$ 90.9 = \dfrac{1}{1+y/2} + \dfrac{1}{(1+y/2)^2} + \dfrac{1}{(1+y/2)^3} + \dfrac{101}{(1+y/2)^4} $$ 

$$ \implies yield = 6,95% $$



#### Conceptos para hacer cobertura

La idea principal detrás de hacer hedging con la duration del bono es la de eliminar el problema multidimensional que plantea las diferentes tasas presentes en el precio de un bono (ver el disclaimer). Para ello, se hace el supuesto simplificador que existe un único factor, la YTM, que afecta al precio del bono y que sirve de aproximación de toda la estructura de riesgo. Notemos que la YTM es un promedio de las tasas, lo que implica que la curva de tasas de interés es plana. Para estudiar la sensibilidad del precio del portafolio utilizaremos una expansión de Taylor del mismo: 

$$ P_t = P(y) = \sum_{i=1}^m \dfrac{C_i}{(1+y)^{T_i-t}} $$ 

$$ dP(y) = P'(y)dy + o(y) $$ 

Donde $P'(y) = - \sum_{i=1}^m \dfrac{(T_i-t)C_i}{(1+y)^{T_i - t + 1}} $ 

Si denominamos la primera derivada P'(y) como $dur (la duration en pesos), podemos ver que el cambios del valor del portafolio puede ser expresado en términos de esta medida: $dP(y) = \$Dur (dy)$ (asumiendo que el residuo de Lagrange es despreciable)

**Definición**: La derivada del precio del portafolio con respecto a su YTM es conocida como la $duration (o sensibilidad): 

$$ \$Dur = P'(y) = - \sum_{i=1}^m \dfrac{(T_i - t)C_i}{(1+y)^{T_i - t + 1}} $$ 

Asimismo, podemos ver que $ \$Dur < 0$, ya que mayores yields están asociados con menores precios (si sube la yield es como si aumentara el factor de descuento de las personas por lo que el costo de oportunidad sería más grande). 

**Definición**: la modifed duration (MD) del portafolio viene dada por: 

$$ MD(P(y)) = - \dfrac{P'(y)}{P(y)} = \sum_{i=1}^m \dfrac{(T_i - t) C_i}{(1+y)^{T_i - t + 1}} \dfrac{1}{P_i} $$ 

Que es una aproximación del cambio proporcional en el precio del portafolio. 

Estas dons medidas son muy útiles, ya que la duration y la modifed duration no solo mide la volatilidad del portafolio, sino que nos permiten computar la ganancia o pérdida (P&L) absoluta o relativa del portafolio antes cambios **pequeños y paralelos en la YTM, $\Delta y $

$$ P\&L \space absoluta \approx \$Dur \Delta y $$

Es decir, si queremos saber cuanto bajará (en unidades de dinero) el bono ante un cambio en la tasa de interés, podemos aproximarlo mediante el cambio en la tasas multiplicado por la duration

$$ P\&L \space relativa \approx -MD \Delta y $$

Es decir, si queremos calcular en que porcentaje caerá el bono ante un cambio en la tasa, podemos aproximarlo de la forma mencionada

Hay que notar que el calculo del cambio en el precio usando la duration es en realidad una aproximación. En realidad, el precio del bono es una función convexa de la tasa interna de retorno, mientras que la aproximación con la duration es una aproximación lineal. Por eso es que decimos que esto sirve para medir cambios "chicos"

#### ¿Cómo hacer el hedging en la práctica? 

Intentaremos hacer el hedging de un portafolio de bonos con $YTM = y$ junto al precio p(y) con otro bono con $YTM = y_1$ (a priori diferente de y), cuyo precio estará dados por $H(y_1)$. Construiremos un portafolio global con precio $P^*$ compuesto por el portafolio P y una proporción $\Phi$ del activo H: 

$$ P^* = P(y) + \Phi H(y_1) $$ 

La idea es hacer que este portafolio global no sea sensible a cambios pequeños en la tasa de interés. Para esto, usaremos la usaremos la aproximación de Lagrange y asumiremos que la curva de la YTM se ve afectada solo por cambios paralelos, por lo que $dy = dy_1$. De esta forma obtenemos: 

$$ dP^* = [P'(y) + \Phi H'(y_1)] dy = 0 $$ 

Entonces, para que el portafolio esté cubierto debe ser cierto que: 

$$ \Phi^* = - \dfrac{P'(y)}{H'(y_1)} = - \dfrac{\$Dur(P(y))}{\$Dur(H(y_1))} = - \dfrac{MD(P(y)) p(y)}{MD(H(y_1)) H(y_1)} $$

Como se puede ver, la cobertura requiere tomar una posición contraría en el activo que se utiliza para hacer el hedging. La idea es que cualquier pérdida (ganancia) obtenida con el portafolio inicial sea cubierta por una ganancia (pérdida) del activo utilizando utilizando para hedgear. En otras palabras, dado que asumimos que la yield curve es plana, usamos la Duration de los distintos portafolios para encontrar la proporción que se debe invertir en el activo utilizado para hacer el hedging. 

Como vimos anteriormente, un supuesto fundamental del análisis nos indicaba que solo podíamos hedgear el riesgo del portafolio para pequeños cambios en la tasa de interés. En casos en que los cambios en la tasa de interés no son pequeños debemos utilizar una expansión de Taylor de segundo orden: 

$$ dP(y) = P'(y)dy + \dfrac{1}{2}P''(y) (dy)^2 $$ 

$$ dP(y) = P'(y)dy + \dfrac{1}{2} \$Conv(P(y)) (dy)^2 $$ 

El termino Conv(P(y)) es conocido con el convexity del portafolio P. 

**Definición**: La convexity del bono se define como: 

$$\$Conv(P(y)) = P''(y) = \sum_{i=1}^m \dfrac{(T_i - t)(T_i - t +1)C_i}{(1+y)^{T_i - t + 2}} $$ 

Como antes, también podemos obtener una aproximación del cambio relativo en el precio del portafolio ante cambios en la tasa de interés: 

$$ \dfrac{dP(y)}{P(y)} \approx -MD(P(y)) dy + \dfrac{1}{2} RC(P(y)) (dy)^2 $$

Donde $RC(P(y)) = \dfrac{P''(y)}{P(y)}$, es conocido como la convexidad relativa del portafolio P. 



#### Como hedgearse ante cambios "grandes" 

La idea es encontrar un portafolio que sea duration-neutral y convexity-neutral. Para ello utilizamos 2 bonos $H_1(y_1)$ y $H_2(y_2)$ para hacer el hedging. Lo que debemos elegir son la cantidades óptimas $\Phi_1$ y $\Pi_2$ de dichos activos. El nuevo portafolio $P^*$ tendrá la siguiente forma: 

$$ P^* = P(y) + \Phi_1 H_1(y_1) dy_1 + \Phi_2 H_2(y_2)dy_2 $$

Las proporciones phi deberán satisfacer: 

- Duration-neutral: $P^{'*} = 0 \implies -P'(y)dy = \Phi_1 H'_1(y_1)dy_1 + \Phi_2 H'_2(y_2) dy_2$

- Convexity neutral: $P^{''*} = 0 \implies -P''(y)(dy)^2 = \Phi_1 H'_1(y_1) (dy_1)^2 + \Phi_2 H'_2(y_2) (dy_2)^2$

Si suponemos cambios paralelos en la yield curve ($dy = dy_1 = dy_2$): 

$$ \begin{cases} -P'(y) = \Phi_1 H'_1(y_1) + \Phi_2 H'_2(y_2) \\ -P''(y) = \Phi_1 H''_1(y_1) + \Phi_2 H''_2(y_2) \end{cases} $$

Solucionando el sistema se obtienen los phis que garantizan hedge ante cambios grandes. 

#### Eliminamos el supuesto de movimientos paralelos 

Hasta aquí hemos analizado como realizar el hedging de un portafolio ante cambios en la yield curve, un caso muy particular. Sin embargo, en la práctica los movimientos de la misma son muy variados habiendo cambios no solo en el nivel, sino en la pendiente y la curvatura que no eran capturados bajo los supuestos anteriormente estudiados. Vemos entonces un modelo general que relaja el supuesto de que los movimientos en la yield curve son paralelos. 

Expresemos el precio $P_t$ del portafolio en el momento t utilizando toda la curva de tasa de zero-coupon: 

$$ P_t = \sum_{i=1}^m \dfrac{C_i}{(1 + R(r,T_i - t))^{T_i - t}} $$

El factor de riesgo en este caso es toda la yield curve, representada a priori por m componentes (en comparación al análisis anterior que tenía uno solo, la YTM). Por simplicidad denotaremos a $R(t, T_i - t)$ como $R_t^i$. COnsideremos por simplicidad cambios pequeños en la Yield curve. Para ello necesitamos solo realizar una expansión de primer orden: 

$$ dP_t \approx \sum_{i=1}^m \dfrac{\partial P_t}{\partial R_t^i} dR_t^i $$ 

Asumamos que el inversor utilizará m =2 activos para hacer el hedging (igual al número de factores de riesgo que es igual al número de madureces) cuyos precios los expresamos como $H^j$. Esta estrategia en la práctica puede resultar muy costosa ya que implica comprar m activos para hacer el hedging (con potencialmente elevados costos de transacción). El precio de estos 2 activos dependerá de todas las tasas de zero-coupon por lo que la expansión de Taylor de primer orden será: 

$$ dH_t^j \approx \dfrac{\partial H_t^j}{\partial R_t^1} d R_t^1 + \dfrac{\partial H_t^j}{\partial R_t^2} dR_t^2 $$ 

Luego construimos nuestro portafolio global $P^*$ como: 

$$ P^* = P_t + \Phi_1 H_t^1 + \Phi_2 H_t^2 $$ 

La condición que deben verificar los phis es: 

$$ dP^* = 0 $$ 

$$ \implies \left(\dfrac{\partial P_t}{\partial R_t^1} + \Phi_1 \dfrac{\partial H_t^1}{\partial R_t^1} + \phi_2 \dfrac{\partial H_t^2}{\partial R_t^1}\right) dR_t^1  +  \left(\dfrac{\partial P_t}{\partial R_t^2} + \Phi_1 \dfrac{\partial H_t^1}{\partial R_t^2} + \phi_2 \dfrac{\partial H_t^2}{\partial R_t^2}\right) dR_t^2 $$

--------------------------------------------

De forma más general podríamos decir que: 

$$ \sum_{i = 1}^m \left(\dfrac{\partial P_t}{\partial R_t^i} + \sum_{j=1}^m \Phi_j \dfrac{\partial H_t^j}{\partial R_t^i}   \right) dR_t^i = 0 $$ 

Donde i representa cada activo y j cada factor de riesgo (ojo con esto). 

--------------------------------------------

Por lo que para hedgear este portafolio de m+1 bonos se debe cumplir para cada i: 

$$ \dfrac{\partial P_t}{\partial R_t^i} + \sum_{j=1}^m \Phi_j \dfrac{\partial H_t^j}{\partial R_t^i} = 0 $$

Notar que si definimos la siguientes matrices: 

$$ H' = \begin{pmatrix} \dfrac{\partial H_t^1}{\partial R_t^1} & \cdots & \dfrac{\partial H_t^m}{\partial R_t^1} \\ \vdots & & \vdots \\ \dfrac{\partial H_t^1}{\partial R_t^m} & \cdots & \dfrac{\partial H_t^m}{\partial R_t^m} \end{pmatrix} $$

$$ \Phi = \begin{pmatrix} \Phi_1 \\ \vdots \\ \Phi_m \end{pmatrix} $$ 

$$ P' = \begin{pmatrix} -\dfrac{\partial P_t}{\partial R_t^1} \\ \vdots \\  -\dfrac{\partial P_t}{\partial R_t^m} \end{pmatrix} $$

El sistema puede ser escrito como: 

$$ H' \Phi = P' $$ 

Donde la solución puede ser escrita como: 

$$ \Phi = (H')^{-1} P' $$ 

Es importante notar que el supuesto que hemos utilizado es que la matriz $H_t$ es invertible, lo que quiere decir que ninguno de los instrumentos utilizados para hacer el hedging es una combinación lineal de los otros m-1 instrumentos. 

----------------------------------------------------

Puede parecer lógico usar componentes principales en este punto. Es de hecho lo ideal para abaratar los costos de transacción. Si se considera, por ejemplo, el caso de 4 tasas de distinta madurez y 2 componentes principales. Si bien hay 4 madureces (con sus correspondientes tasas), hay dos componentes principales que explican la mayor parte de la variabilidad de dichas tasas. Por lo tanto, utilizo 2 bonos para hedgear: **si bien hay 4 tasas, son solo 2 las fuentes de riesgo (los dos componentes principales)**
