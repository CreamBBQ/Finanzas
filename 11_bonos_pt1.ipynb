#### Introducción y algunas definiciones 

En esta sección se trabajará con instrumentos financieros de ingreso fijo. Estos instrumentos tienen la peculiaridad que al ser comprados uno sabe cual es el rendimiento de dichos instrumentos. Un bono es uno de los instrumentos más simples de ingreso fijo. 

Los bonos pueden pagar o no cupones desde el momento que se compran hasta el momento del vencimiento del bono. Los bonos ás simples de analizar son aquellos que no pagan cupones. El problema clave es determinar el precio justo a pagar por un bono que paga al momento del vencimiento un cantidad determinada. Decimos que un *zero coupon bond* es un bono que desde el momento de su compra hasta su vencimiento no paga cupones. 

Comenzaremos dando definiciones sobre distintos elementos que componen el valor de un bono asumiendo certidumbre. Vamos a ver más adelante que el paso de un mundo de certidumbre a uno de incertidumbre es el que plantea la problemática de hallar el precio justo del bono. 

Si se supone que el bono paga 1 dólar en T, el momento del vencimiento, entonces de denota el precio del bono a tiempo t como: P(t). Denotando a la tasa de rendimiento del bono como R(t,T) se define que el precio de lobo a tiempo t debe ser el valor descontado (por la tasa de retorno) de un peso. Como no hay incertidumbre, el valor de recibir ese dólar en el futuro tiene que ser igual l valor presente de ese dólar hoy. Ahora bien, para calcular el valor actual de ese dólar a recibir en el futuro en términos de dólares actuales, lo único que hay que hacer es descontarlo mediante la tasa de interés. Como trabajamos en tiempo continuo, para descontar hay que multiplicar por $e^{-R(t,T)(T-t)} * 1$. Por no arbitraje, ese tiene que ser el precio del bono, definido entonces como: 

$$ P(t,T) = e^{-R(t,T)(T-t)} $$

Por lo tanto, la relación entre el precio y la tasa de retorno de un zero coupon bond es: 

$$ R(t,T) = - \dfrac{1}{T-t}ln[P(t,T)] $$


Se define una tasa forward como aquella tasa que es conocida en t, pero que se aplica entre $S_1$ y $S_2$, donde $t< S_1 < S_2$. Usando la definición anterior definimos el precio del bono en el periodo t igual al calor descontado (a la tasa de rendimiento) del precio del bono en el futuro. En este caso particular la tasa forward se aplica al rendimiento del bono en el intervalo $[S_1, S_2]$, por ende: 

$$ P(t, S_2) = P(t,S_1)e^{-r(t,S_1,S_2)(S_2 - S_1)} $$

La tasa instantánea forward se define como el limite de la tasa forward cuando $S_2 \to S_1$, en otras palabras la tasa de rendimiento contratada a tiempo t para el periodo $[S_1, S_1 + dt]$.

Usando que $\lim_{x \to 0}  e^x = \lim_{x \to 0} (1+x)$ tenemos

$$ \lim_{S_2 \to S_1} e^{-r(t,S_1,S_2)(S_2 - S_1)} = \lim_{S_1 \to S_2} (1- r(t,S_1,S_2)(S_2-S_1)) $$

Podemos computar el límite para $P(t,S_2)$ cuando $S_2 \to S_1$ 

$$ \lim_{S_2 \to S_1} P(t, S_2) = P(t, S_1) \lim_{S_2 \to S_1} (1- r(t,S_1,S_2)(S_2-S_1)) $$ 

$$ \lim_{S_2 \to S_1}\dfrac{P(t, S_2) }{P(t, S_1) }= \lim_{S_2 \to S_1} (1- r(t,S_1,S_2)(S_2-S_1)) $$ 

$$ \lim_{S_2 \to S_1}\dfrac{P(t, S_2) }{P(t, S_1)} - 1 = r(t, S_1) \lim_{S_2 \to S_1} (S_2 \to S_1) $$ 

$$ \lim_{S_2 \to S_1}\dfrac{P(t, S_2) }{P(t, S_1) (S_2 - S_1)} - \dfrac{P(t, S_1)}{(S_2 - S_1)P(t, S_1)} = - r(t, S_1) $$ 

$$ \lim_{S_2 \to S_1} \dfrac{P(t, S_2) - P(t, S_1)}{P(t,S_1)(S_2 - S_1)} = -r(t, S_1) $$

Notar que es el producto de una constante por el límite del cociente incremental, es decir, la derivada del precio con respecto a $S_1$. Reordenando se obtiene la tasa instantánea forward. 

$$ r(t, S_1) = \dfrac{-1}{P(t,S_1)} \dfrac{dP(t,S_1)}{dS_1} $$ 

Usando esta ecuación podemos definir la tasa spot (instantánea) en el momento t como el límite de la tasa instantánea forward cuando $S_1$ tiende a t (como la tasa instantánea futura cuando "el futuro" tiende a cero), por lo tanto, 

$$ R(t,t) = r(t,t) = r(t) $$ 

De esta manera podemos escribir el precio de un bono como 1 peso descontando por la suma de tasas forward entre t y T (notar que a tiempo t se conocen todas las tasas instantáneas forward)

$$ P(t,T) = e^{-\int_{t}^{T} r(t,u)du} $$

Además, podemos reemplazar en la ecuación del retorno y obtener

$$ R(t, T) = \dfrac{1}{T-t} \int_{t}^{T} r(t,u)du $$ 

Esta conclusión es importante: **el retorno es igual a un promedio de tasas forward que son conocidas a tiempo t y contratadas para tiempo** $ u \in [t,T]$

Si se supone que no existe incertidumbre, entonces la tasa spot en el período S debe ser, por el principio de no arbitraje, igual a la tasa instantánea forward en t contratada para S, es decir, $r(S) = r(t,S)$. Entonces bajo ese supuesto, el retorno del bono no es más que un promedio de las tasas instantáneas spot entre t y T. Esto permite conocer la evolución del retorno o yield curve en cualquier momento del tiempo a través del valor de la tasa r (la yield curve es una función que relaciona el retorno de bonos de distinta madurez con la madurez para un momento del tiempo). Notemos que la tasa futura de retorno instantánea bajo certidumbre la asumimos como conocida, sin embargo, en un mundo con incertidumbre esto no será cierto. Como introducir un modelo de predicción de las tasas spot en el futuro es una de las fuentes de mayor confusión al hacer el pricing de los bonos. Por lo tanto, al introducir incertidumbre sobre la tasa instantánea spot de interés en el futuro, vamos a asumir que sigue un proceso estocástico determinado. De este modo vamos a poder obtener como resultante el precio de un bono y el precio de las tasas de interés a futuro. 




#### One - Factor interés rate modelling 

COmo la tasa de interés no es conocida en el futuro, se supone que r sigue el siguiente proceso estocástico: 

$$ dr = u(r,t) dt + w(r,t)dz $$ 

Vamos a caracterizar posibles modelos (de un factor) para obtener el precio de los bonos mediante distintos supuestos sobre u(r,t) y w(r,t). El pricing de bonos lo hacemos construyendo un portafolio que consiste de comprar y vender bonos (que son funciones de la tasa instantánea de interés) de distintas madureces. Esto lo hacemos de este modo pues de puede demostrar que no es posible construir un portafolio entre un bono de madurez T y la tasa instantánea de interés (un bono de esa madurez) y eliminarle la incertidumbre a dicho portafolio. Este problema es análogo al que se nos presentaba cuando hacíamos opciones sobre una variable no comerciable como la tasa de inflación. 

------------------------------------------------------------------

Demostración de por que no podemos hacer hedging con un bono de vencimiento instantáneo: 

Intuitivamente uno pensaría que se puede formar el portafolio haciendo hedging entre una cartera que consiste en un bono de madurez T y un bono instantáneo. Consideremos así la posibilidad de realizar un cartera que consiste de un bono de madurez $T_1$ y uno de madurez t + dt. Para realizar el pricing procederíamos a eliminar la incertidumbre de dicho portafolio e igualar dicho importe a lo que se obtendría a la tasa libre de riesgo, 

$$ d\Pi = W_1 dP(r,t,T_1) - W_2dP(r,t, t+dt) $$

Luego si aplicamos el lema de Ito para ver el cambio del precio del bono obtenemos: 

$$ dP = P_t dt + P_r (u(r,t)dt + w(r,t)dz) + \dfrac{1}{2}w(r,t)^2P_{rr}dt $$

Ahora veremos que debido a las características del bono instantáneo no vamos a poder eliminar la incertidumbre del portafolio. Recordemos que por definición 

$$ P(t, t + dt) = e^{-r(t)dt} $$ 

Donde tomando límites cuando dt tiende a cero obtenemos que 

$$ P(t, t + dt) \approx 1 - r(t) dt $$

De la ecuación anterior se desprende que: 

$$ P_r = -dt $$

$$ P_{rr} = 0 $$ 

Entonces el cambio en P va ser de la siguiente manera: 

$$ dP = P_t dt + u(r,t)dt^2 + w(r,t)dz(-dt) + \dfrac{1}{2}w(r,t)^2(0)dt $$

Recordando los términos despreciables que se van rápidamente a cero concluimos que: 

$$ dP = P_t dt  $$ 

Y por lo tanto, al no tener ningún termino asociado al riesgo, no se va a poder eliminar la incertidumbre del portafolio propuesto, intuitivamente esto prueba que como el precio de un bono de madurez instantánea es inmediatamente 1, no hay posibilidad de realizar hedging. Vamos a tener que calcular el precio del bono utilizando un portafolio con bonos de distinta madurez, y por lo tanto el resultado va a ser una función del precio del riesgo (como pasaba en el caso 3 de pricing).

#### Valuación del precio de los bonos 

La característica principal del pricing de bonos es que no podemos armar una cartera comprando o vendiendo el bono instantáneo, por lo tanto se realiza el pricing utilizando 2 bonos de distinta madurez, $T_1$ y $T_2$. Para encontrar el precio del bono construimos un portafolio $P_1(r,t,T_1)$ y $P_2(r,t,T_2)$ donde el retorno del portafolio está definido como: 

$$ \dfrac{d \Pi}{\Pi} = \alpha \dfrac{d P_1}{P_1} + (1-\alpha) \dfrac{d P_2}{P_2} $$ 

y 

$$ \dfrac{d P_i}{P_i} = \mu_i dt + \sigma_i dz $$ 

Por lo tanto, el cambio del portafolio es igual a: 

$$ \dfrac{d \Pi}{\Pi} = [ \alpha \mu_1 + (1-\alpha) \mu_2 ]dt + (\alpha \sigma_1 + (1-\alpha)\sigma_2) dz $$

Para eliminar el riesgo se elige un $\alpha$ tal que el termino que acompaña a dz sea nulo. De esta forma se tiene que: 

$$ \alpha = \dfrac{\sigma_2}{\sigma_2 - \sigma_1} $$ 

Una vez eliminado el riesgo, por el principio de no arbitraje, el retorno del portafolio debe ser igual a la tasa libre de riesgo: 

$$ \dfrac{d \Pi}{\Pi} = r dt $$ 

$$ [ \alpha \mu_1 + (1-\alpha) \mu_2 ]dt = r dt $$ 

$$ \left[ \dfrac{\sigma_2}{\sigma_2 - \sigma_1}  \mu_1 + \left(1 - \dfrac{\sigma_2}{\sigma_2 - \sigma_1}  \right) \mu_2 \right]dt = r dt $$ 

$$ \implies \dfrac{\mu_1 - r}{\sigma_1} = \dfrac{\mu_2 - r}{\sigma_2} = \lambda $$ 

Donde lambda es el precio del riesgo de cada bono. POr el lema de Ito sabemos que: 

$$ \sigma_i = \dfrac{1}{P_i} w(r,t)P_{ir} $$ 

Además, 

$$ \mu_i = \dfrac{1}{P_i} \left( P_{it} + P_{ir} u(r,t) + \dfrac{1}{2} P_{irr} w(r,t)^2 \right) $$

Reemplazando en la igualación de precios del riesgo y reordenando se obtiene que: 

$$ P_{it} + (u(r,t) - \lambda w(r,t) ) P_{ir} + \dfrac{1}{2} P_{irr} w(r,t)^2 - P_i r = 0 $$

Obtenemos una ecuación diferencial que depende de $\lambda$ pues no podemos hacer hedging con el subyacente y de ese modo eliminar el precio del riesgo. Aquí lambda debe ser obtenido independientemente. 

A continuación se presentan dos modelos para hacer el pricing. Cada uno de ellos propone un proceso diferente para calcular la tasa spor modificando el supuesto sobre que proceso estocástico sigue la tasa instantánea de interés, o sea, modificando las funciones $u(r,t)$ y $w(r,t)$. 

Es importante aclarar que cuando realizamos el pricing de los bonos no tenemos ninguna razón para sospechas (como con las acciones que por eficiencia a la Fama suponíamos que seguían un Random Walk) que dicha tasa sigue un proceso estocástico determinado. Es probable que en algunos momentos del tiempo sea mejor caracterizada por uno u otro proceso estocástico. Será un problema de carácter empírico decidir cual supuesto sobre la evolución de las tasas de corto plazo se condice con una muestra determinada. A continuación pasaremos a encontrar el precio de los bonos bajo el supuesto que la tasa instantánea sigue deferentes procesos estocásticos. 

#### Modelos de Merton 

Merton propone el proceso más simple para la tasa spot donde, $u(r,t) = \alpha$ y $w(r,t) = \sigma$, y por lo tanto la tasa de interés sigue una BMA: 

$$ dr = \alpha dt + \sigma dz $$

Sustituyendo en la solución general obtenemos, 

$$ P_t + (\alpha - \lambda \sigma) P_r + \dfrac{1}{2} \sigma^2 P_{rr} - r P = 0 $$ 

La ecuación anterior es diferencial, por lo cual debemos (i) encontrar un solución general, (ii) imponer una condición terminal. La condición terminal corresponde al pay-off del bono de madurez, es decir, $P(r,T,T) = 1$. La solución general de este tipo de ecuaciones diferenciales parciales es: 

$$ P(r,t,T) = e^{-r\tau + F(\tau)} $$ 

Donde $\tau = T-t$ y F es una función del tiempo a determinar. La solución por definición sabe satisfacer la ecuación diferencial parcial por lo que evaluaremos sus derivada y sustituiremos en dicha ecuación. Esto nos permitirá determinar F y obtener la solución para los bonos en forma explícita. 

$$ P_r = -\tau e^{-r\tau + F(\tau)} $$ 

$$ P_{rr} = \tau^2 e^{-r\tau + F(\tau)} $$ 

$$ p_t = (r - F'(\tau))e^{-r\tau + F(\tau)} $$ 

Sustituyendo en la ecuación diferencial obtenemos: 

$$ \left( r- F'(\tau) - (\alpha - \lambda \sigma) \tau + \dfrac{1}{2} \sigma^2 \tau^2 - r\right) e^{-r\tau + F(\tau)} = 0 $$ 

Por lo que debe ser cierto que: 

$$ F'(\tau) = -(\alpha - \lambda \sigma) \tau + \dfrac{1}{2} \sigma^2 \tau^2 $$ 

O bien, 

$$ F(\tau) - (\alpha - \lambda \sigma) \dfrac{\tau^2}{2} + \dfrac{1}{6} \sigma^2 \tau^3 + C $$

Sustituyendo el precio del bono y usando la condición terminal de que $P(r,T,T) = 1$, obtenemos que C = 0, por lo que la solución para el precio del bono puede ser escrita como: 

$$ P(r,\tau) = exp\{-r\tau - \frac{1}{2}(\alpha - \lambda \sigma) \tau^2 + \frac{1}{6}\sigma^2 \tau^3 \} $$

Por lo que el retorno es igual a: 

$$ R(r,T) = -r\tau - \frac{1}{2}(\alpha - \lambda \sigma) \tau^2 + \frac{1}{6}\sigma^2 \tau^3  $$

Notar que si $\tau \to \infty$ entonces la tasa de retorno que estaría ofreciendo el bono sería negativa. Por lo tanto el modelo de Merton no debería en principio usarse para bonos de muy larga madurez. Una vez conocido el precio del bono podemos fácilmente calcular la tasa forward como: 

$$ r(t,T) = \dfrac{-1}{P(t,T)} \dfrac{dP(r,T)}{dT} = r + (\alpha - \lambda \sigma) \tau - \dfrac{1}{2} \sigma^2 \tau^2 $$
 


#### Modelo de Vasicek 

Para evitar un yield negativo en el largo plazo se puede proponer un modelo alternativo para la tasa spot, en vez de suponer que r sigue un proceso Brownian Motion, Vasicek propuso un proceso de Mean Reverting Brownian Motion. Ahora el proceso de la tasa spot r es: 

$$ dr = u(r,t) dt + w(r,t) dz $$

Con $ u(r,t) = \alpha (\gamma - r)$ y $w(r,t) = \sigma$

Bajo este proceso estocástico las tasa sport va a fluctuar alrededor de una media $\gamma$. Si la tasa r baja (con respecto a la de largo plazo), entonces el termino $\alpha(\gamma-r)dt$ es positivo, por lo que es muy probable que la tasa de interés suba en el futuro. El proceso de r se puede ver como un proceso estocástico en el que eventualmente la tasa de interés retorna a su media (o su valor largo plazo). El escalar $\alpha$ define con que frecuencia la tasa retorna a la media. Si es pequeño la tasa no retornará muy rápido (o seguido) a la media, pero si es grande la frecuencia de retorno es mayor. Por analogía la nueva ecuación diferencial es igual a: 

$$ P_t + \dfrac{1}{2}\sigma^2 P_{rr} dt + (\alpha(\gamma-r) - \lambda \sigma) P_r - r P = 0 $$ 

Definimos ahora $\hat{\gamma} = (\gamma - \frac{\lambda}{\alpha}\sigma) $ De forma tal que la expresión de interés es: 

$$ P_t + \dfrac{1}{2} \sigma^2 P_{rr} dt + \alpha(\hat{\gamma} - r) P_r - rP = 0$$

La solución (Guess) que se propone es: 

$$ P = exp \{ A(\tau) + B(\tau) r \} $$ 

El procedimiento para encontrar la solución es igual al que mostramos en detalle para el modelo de Merton. Tras resolver la ecuación se encuentran las siguientes relaciones: 

$$ A(\tau) = -\dfrac{1 - exp \{-\alpha \tau \}}{\alpha} $$

$$ B(\tau) = \left[ \dfrac{(-B(\tau) - \tau)(\alpha^2 \hat{\gamma} - \frac{1}{2} \sigma^2)}{\alpha^2} \right] - \dfrac{1}{4\alpha}\sigma^2 B(\tau)^2 $$

Siguiendo el mismo razonamiento del modelo anterior, se obtiene que el yield del bono es igual a: 

$$ R(t, T) = -\dfrac{1}{\tau} [ln(P(t,T))] = -\dfrac{1}{\tau} [A(\tau) + B(\tau) r] $$

Dada esta expresión se concluye que cuando $\tau$ es muy grande $R(t,T)$ tiende a $\hat{\gamma} - \frac{1}{2}(\frac{\sigma}{\alpha})^2 $, que para valores razonables uno esperaría que sea positivo

#### Two-factor interest rate modelling 

Modelos simples de un factor como los presentados en la sección anterior no siempre pueden describir el comportamiento de la yield curve. EL problema es que en ese tipo de modelos se usa solo información de la tasa de interés instantánea para predecir todo el yield curve. En este sección se presentará un modelo que propone que el valor del bono depende de la tasa de interés spot y de un bono de madurez infinita. De este modo se busca utilizar información sobre los 2 extremos para predecir el comportamiento del yield curve, o hacer el pricing del bono.

Considere el precio de un bono con madurez T como una función de estas 2 tasas de retorno, $P(r,l,t,T)$, donde l es la tasa de retorno del bono de madurez infinita. La evolución de estas tasas es la siguiente: 

$$ dt = \mu_r dt +  \sigma_r dz_r $$ 

$$ dl = \mu_l dt +  \sigma_l dz_l$$ 

Donde $ E[dz_r dz_l] = \rho dt $ 

En general se define a l como una tasa de interés de largo plazo, el particular se fija l como el retorno de un Bono Consol, que es un bono de madurez infinita que paga cupones. Luego, el precio de un bono consol que paga cupones c por período es: 

$$ C_o = \int_0^\infty c e^{-lt} dt = \dfrac{c}{l} $$

Para encontrar el precio del derivado consideramos un portafolio que consiste en 3 bonos de distinto vencimiento $T_1$, $T_2$ y $T_3$. El retorno del portafolio es: 

$$ \dfrac{d \Pi}{\Pi} = \alpha_1 \dfrac{d P_1(r,l,t,T_1)}{P_1(r,l,t,T_1)} + \alpha_2 \dfrac{d P_2(r,l,t,T_2)}{P_2(r,l,t,T_2)} + \alpha_3 \dfrac{d P_3(r,l,t,T_3)}{P_3(r,l,t,T_3)} $$ 

Donde la sumatoria de los alphas debe ser igual a uno. Por el lema de Ito sabemos que el cambio proporcional en el precio de los bonos sigue un movimiento browniano donde la incertidumbre de dicho cambio es una función de $dz_l$ y $dz_r$, por lo tanto: 

$$ \dfrac{d P_i(r,l,t,T_i)}{P_i(r,l,t,T_i)} = \mu_i dt + \sigma_{il} dz_l + \sigma_{ir} dz_r $$

Por lo tanto el retorno del portafolio puede ser escrito como: 

$$ \dfrac{d \Pi}{\Pi} = (\alpha_1 \mu_1 + \alpha_2 \mu_2 + \alpha_3 \mu_e) dt + (\alpha_1 \sigma_{1l} + \alpha_2 \sigma_{2l} + \alpha_3 \sigma_{3l}) dz_l + (\alpha_1 \sigma_{1r} + \alpha_2 \sigma_{2r} + \alpha_3 \sigma_{3r}) dz_r $$

Una vez que elijamos los alphas que eliminan el riesgo, el retorno de dicho portafolio debe ser igual a la tasa libre de riesgo. Esto me genera el siguiente sistema: 

$$ \begin{pmatrix} \sigma_{1l} & \sigma_{2l} & \sigma_{3l} \\ \sigma_{1r} & \sigma_{2r} & \sigma_{3r} \\ \mu_1 - r & \mu_2 - r & \mu_3 - r \end{pmatrix}
\begin{pmatrix} \alpha_1 \\ \alpha_2 \\ \alpha_3 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \\ 0 \end{pmatrix} $$

Este sistema tiene la peculiaridad de que permite expresar cualquier fila como una combinación lineal de las otras (ya que los alpha suman 1, por lo que encontramos una solución al sistema homogéneo que no es cero), y por lo tanto, encontrar el exceso de retorno del bono como: 

$$ \mu_P - r = \lambda_r \sigma_r + \lambda_l \sigma_l $$ 

Notar que los lambdas son el precio de mercado del riesgo de cada factor y es independiente de la madurez de cualquier bono (ver APT). Si aplicamos lema de ito tenemos que: 

$$ dP(r,l,t) = P_r dr + P_l dl + \dfrac{1}{2}P_{rr}(dr)^2 + \dfrac{1}{2} P_{ll} (dl)^2 + P_{rl} dr dl $$ 

$$ dP(r,l,t) = (P_t + P_r \mu_r + P_l \mu_l + \dfrac{1}{2} P_{rr} \sigma_r^2 + \dfrac{1}{2} P_{ll} \sigma_l^2 + P_{rl} \rho \sigma_r \sigma_l) dt + P_r \sigma_r dz_r + P_l \sigma_l dz_l $$

Si asumimos que el bono consol paga cupones cdt por unidad de tiempo este retorno debe ser sumado a la media de los bonos.

$$  \dfrac{dP(r,l,t)}{P(r,l,t)} = \dfrac{1}{P}(P_t + P_r \mu_r + P_l \mu_l + \dfrac{1}{2} P_{rr} \sigma_r^2 + \dfrac{1}{2} P_{ll} \sigma_l^2 + P_{rl} \rho \sigma_r \sigma_l + c) dt +  \dfrac{1}{P} P_r \sigma_r dz_r +  \dfrac{1}{P} P_l \sigma_l dz_l $$

Reemplazando en la ecuación de no arbitraje se obtiene: 

$$ P_t + \dfrac{1}{2} \sigma_r^2 P_{rr} + \dfrac{1}{2} \sigma_l^2 P_{ll} + \rho \sigma_r \sigma_l P_{rl} + (\mu_r - \lambda_r \sigma_r) P_r + (\mu_l - \lambda_l \sigma_l) P_l - rP +c = 0$$

Esta ecuación tiene a los lambdas como incógnitas por lo que en principio habría que obtener los dons precios con información independiente. Sin embargo, podemos explotar el hecho de que sabemos cual es la forma estructural del precio de los consols para re escribir esta ecuación y eliminar un precio del riesgo. En particular dicha ecuación diferencial debe ser valida para todo bono incluyendo los consols, por lo tanto, sin reemplazamos las derivadas de dicha ecuación, encontramos un precio del riesgo como función de las variables conocidas (recordar que P = \frac{c}{l})

$$ \implies (\mu_l - \lambda_l \sigma_l) = \dfrac{\sigma_l^2}{l} + l^2 - rl  $$

Reemplazando en la ecuación de interés nos queda todo solo dependiendo del precio del riesgo asociada a la tasa spot: 

$$ P_t + \dfrac{1}{2} \sigma_r^2 P_{rr} + \dfrac{1}{2} \sigma_l^2 P_{ll} + \rho \sigma_r \sigma_l P_{rl} + (\mu_r - \lambda_r \sigma_r) P_r + (\dfrac{\sigma_l^2}{l} + l^2 - rl )P_l - rP +c = 0$$

La prima de riesgo del mercado de la tasa spot no se puede reemplazar y por lo tanto ella debe ser estimada.
