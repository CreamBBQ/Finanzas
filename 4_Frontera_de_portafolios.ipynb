<CENTER> <h1> LA FRONTERA DE PORTAFOLIOS </h1> <center>

En caso de que exista más de un activo riesgoso de modo que el individuo pueda formar distintos portafolios (sin restricciones), si alguno de estos portafolios domina estocásticamente en segundo orden a todo el resto (dada una igual media), entonces este portafolio tiene que ser el que tiene mínima varianza. El conjunto de todos los portafolios que minimizan la varianza (o el desvio estándar), dadas diferentes medias, forman un conjunto que se denomina como la frontera (de portafolios con dicha caracteristica). 

Lo anterior sugiere la existencia de un isomorfismo entre la maximización de la utilidad esperada y la minimización de la varianza (dada una media). Sin embargo, la aversión a una mayor varianza está implicita en la concavidad estricta de la función de utilidad y por lo tanto dicho isomorfismo (entre el modelo de media-varianza y la maximización de la utildad esperada) no es válido para funciones de utilidad y/o distribuciones arbitrarias.

La relación entre el modelo de media-varianza y la maximización de utilida esperada será equivalente en casos puntuales como una función de utilidad cuadrática con la problemática de que esta presenta la propiedad de saciedad y de aversión absoluta al riesgo, lo que implicaría que el riesgo es tratado como un bien inferior. Por otro lado, la propiedad también es valida si la distribución de los retornos es normal, que no caracterizan el comportamiento real de los retornos (se permiten, por ejemplos, retornos infinitamente negativos). 

Por estas razones el modelo de media-varianza no es un modelo general de elección de activos. Pero dado que tiene buenas predicciones empíricas y que, principalmente, es analíticamente manejable, ocupa un rol central en la teoría de elección de portafolios. 


#### Modelos de media-varianza: 

##### Supuestos

- Existen $n\geq 2$ activos riesgosos en la economía
- No existen fricciones en la economía 
- Se permite ir corto 
- Los activos tienen varianzas finitas y diferentes medias 
- El retorno de cualquier activo no puede ser expresado como combinación líneal de retornos en otros activos
- No hay activos libres de riesgo (al menos en principio)

##### Notación

- $r_{nx1} \equiv$ vector de retornos de los n activos riesgosos
- $Var[r]=v_{nxn} \equiv$ matriz de varianzas y covarianzas
- $E[r]=E_{nx1} \equiv$ Media de los retornos
- $rp = \alpha'r \space \space \space$ donde $\alpha$ es el vector de proporción invertida en cada activo 
- $E[\alpha' r]= \alpha'e$
- $Var[\alpha' r]= \alpha'v\alpha$

##### El modelo

Un portafolio es un portafolio frontera si tiene la mínima varianza entre todos los portafolios que tienen su misma media. Un portafolio p es frontera es frontera si y solo si $\alpha_p$ es solución del seiguiente problema: 

$\displaystyle{\min_{w} \dfrac{\alpha'v\alpha}{2} \space \space \space \text{s.a} \space \space \space \alpha'e=E[r_p] \space \space \space \text{y} \space \space \space \alpha'I=1}$

Donde I es in vector de unos (de n-componentes). Notemos que no existe restricción alguna sobre el límite de entradas en corto que puede hacer el individuo. El lagrangiano que resuelve el problema de minimzación es: 

$\displaystyle{\min_{\alpha, \lambda,\gamma} L =  \dfrac{\alpha'v\alpha}{2} + \lambda(E[r_p]-\alpha'e ) + \gamma(1-\alpha'I)}$ 

Recordemos las regulas básicas de derivación de matrices: 

![der.png](https://i.imgur.com/JESWwfi.png)




De este modo obtenemos las condiciones de primer orden como sigue: 

$L_\alpha = (v + v')\dfrac{\alpha}{2} - \lambda e = 0 - \gamma I$

Recordemos que v, la matriz de varianzas y covarianzas, es simétrica, por lo que 

$L_\alpha = v\alpha - \lambda e - \gamma I= 0$

$L_\lambda = E[rp]-\alpha'e=0$

$L_\gamma = 1 - \alpha ' I = 0$

Como V es una matriz definida positiva, estas condiciones son necesarias y suficientes para la existencia de un mínimo global. 

Despejamos $\alpha$ de la primera expresión: 


$\alpha = \lambda v^{-1}  e +  \gamma v^{-1}  I$

Usamos la segunda CPO y el resultado anterior: 

$E[rp]=\alpha'e=e'\alpha= \lambda(e' v^{-1} e) + \gamma (e' v^{-1} I)$

Repetimos el proceso para la tercera condición de primer orden: 

$1=\alpha'I=I'\alpha= \lambda(I' v^{-1} e) + \gamma (I' v^{-1} I)$

Definiendo: 

- $A = I'v^{-1}e = A'$
- $B = e'v^{-1}e$
- $C = I'v^{-1}I$
- $D = BC - A^2$

Entonces se tiene que: 

- $E[rp] = \gamma A + \lambda B$
- $1 = \gamma C + \lambda A$

O, matricialmente, 

$ \begin{bmatrix} B & A \\ A & C \end{bmatrix}
\begin{bmatrix} \lambda \\ \gamma \end{bmatrix} = 
\begin{bmatrix} E[rp] \\ 1 \end{bmatrix} $

Para resolver esto será necesario recordar la regla de Crammer. A continuación de proverán dos ejemplos para sistemas de 2x2 y de 3x3: 

![crammer1.png](https://i.imgur.com/TlWr0uU.png)

![crammer2.png](https://i.imgur.com/snUqyBp.png)


Primero notar que en determinante de la matriz principal es, no por coincidencia, D. Teniendo en cuenta esto y la regla de cramer la solución será: 

$\lambda = \dfrac{\begin{vmatrix} E[rp] & A \\ 1 & C \end{vmatrix}}{D}$

$\implies \lambda = \dfrac{E[rp]C-A}{D}$

$\gamma = \dfrac{\begin{vmatrix} B & E[rp] \\ A & 1 \end{vmatrix}}{D}$

$\implies \gamma = \dfrac{B-E[rp]A}{D}$

Usando los despejes de las condiciones de primer orden se llega a la solución de los pesos óptpimos: 

$\alpha = \dfrac{E[rp]C-A}{D} v^{-1}  e +  \dfrac{B-E[rp]A}{D} v^{-1}  I$

$\alpha = \dfrac{E[rp]Cv^{-1} e}{D} - \dfrac{Av^{-1}  e}{D} + \dfrac{Bv^{-1}  I}{D} - \dfrac{E[rp]Av^{-1}  I}{D}$

$\alpha = E[rp] \underbrace{\left( \dfrac{Cv^{-1} e}{D} - \dfrac{Av^{-1} I}{D}\right)}_{\equiv h} -  \underbrace{\dfrac{Av^{-1}  e}{D} + \dfrac{Bv^{-1}  I}{D}}_{\equiv g}$

$\alpha = E[rp] h + g$

De este modo, un portafolio es frontera si puede ser representado por esta última ecuación. El conjunto de todos los portafolios frontera forman la frontera de portafolio. 

A continuación se presentan una serie de teoremas (cuatro) que serán útiles para definir la forma de esta frontera y sus puntos críticos.

##### Teorema 1 
Si dos portafolios pertenecen al conjunto frontera ($\alpha_{p1}, \alpha_{p2} \in Front(\alpha)$) toda combinación lineal entre ellos también pertenece a la frontera. 
Este teorema nos permite generar toda la frontera a partir de dos portafolios frontera juntos. 

##### Teorema 2 
La covarianza de dos portafolios frontera p y 1 es: 

$Cov(r_p, r_q) = \alpha_p'v\alpha_q=\dfrac{C}{D} \left( E[r_p] - \dfrac{A}{C} \right )  \left( E[r_q] - \dfrac{A}{C} \right )  + \dfrac{1}{C}$





#### Derivación de la frontera de activos 

De la fórmula para la covarianza (recordar que la covarianza entre una variable una aleatoria contra ella misma es igual su varianza) se puede inferir la siguiente expresión para la varianza: 

$\sigma^2(r_p)=\dfrac{1}{C} + \left (E[r_p]- \dfrac{A}{C} \right) ^2 \dfrac{C}{D}$

Reordenando: 

$\dfrac{\sigma^2(r_p)}{1/C} - \dfrac{(E[r_p]-\frac{A}{C})^2}{D/C^2} = 1$

Esta última expresión es la ecuación (implícita) de una hiperbóla en el plano $(\sigma(r_p), E[r_P])$, con centro en $(0, \frac{A}{C})$ y asíntotas $E[r_p] = \dfrac{A}{C} \pm \sqrt{\dfrac{D}{C}}\sigma(r_p)$

De modo que la forntera de portafolios queda de la siguiente manera: 

![frontera.png](https://i.imgur.com/U5L8zC5.png)



##### Portafolio de mínima varianza 

Se puede ver claramente que la mínima varianza se obtiene cuando $E[r_p] = A/C$. El desvío estándar de dicho portafolio es $\sigma(r_p) = \sqrt{1/C}$. Se denota al retorno del portafolio de mínima varianza como $r_{mv}$. 

Se puede demostrar que la covarianza entre el retorno de cualquier portafolio p (no necesariamente de la frontera) y el retorno del portafolio de mínima varianza es siempre igual a la varianza del retorno del portafolio de mínima varianza, es decir 

$Cov(r_p, r_{mv}) = Var(r_{mv})$

##### Portafolios eficientes 

Todos los portafolios frontera que tengan un retorno esperado más alto que el portafolio de mínima varianza, son portafolios eficientes. Debido a la forma de la frontera para cada portafolio eficiente existe un portafolio ineficiente con una misma varianza pero un menor retorno esperado. 

Se puede demostrar que cualquier conbincación convexa de portafolios eficientes será un portafolio eficiente. 

##### Teorema 3 

Si los agentes son estrictamente aversos al riesgo, entonces eligirán portafolios eficientes.

Bajo el supuesto de funciones de utilidad crecientes, los individuos que son aversos al riesgo ven a la varianza del activo, en este caso representado por su desvio estándar, como un mal. Sus curvas de indiferencia sobre la elección de bienes tiene entonces la siguiente forma: 

![ci.png](https://i.imgur.com/XP2Lmcf.png)

Vemos entonces como las curvas de indiferencia son tangentes en la parte eficiente de la frontera. 

##### Portafolio de cero covarianza 

Una propiedad de los portafolios frontera es que para cualquier portafolio p, exceptuando el de mínima varianza, existe un único portaforlio fontera tal que tiene cero convarianza con p, y se denota como zc(p). Usando la formula de la covarianza se obtiene que: 

$E[r_{zc(p)}] = \dfrac{A}{C} - \dfrac{\dfrac{D}{c^2}}{E[r_p] - \dfrac{A}{C}}$

Notar que la solución a la ecuación es única y, además, para que no suceda un absurdo en el último término a la derecha de la ecuación, el portafolio de mínima varianza no puede tener su correspondiente portafolio de cero covarianza. 

Se puede ver, además, que si p es un portafolio eficiente entonces el último término a la derecha de la ecuación es positivo ($E[r_p]>A/C). Se concluye entonces que para cualquier porfatolio frontera eficiente, el correspondiente portafolio de cero covarianza es ineficiente. 


##### Teorema 4 

El portafolio $zc(p) = q$ es una combinación lineal entre el portafolio p y el de mínima varianza, es decir, 

$\alpha_q = \Theta \alpha_p + (1-\Theta) \alpha_{mv}$

Un cololario importante de este teorema es que el portafolio q, de cero covarianza con p, se construye haciendo short en el portafolio p y comprando el de mínima varianza. 

##### Retorno requerido por cualquier activo para ser incluido en un portafolio 

Una de las preguntas más importantes de un individuo que desea comprar un activo riesgoso es: cual es el exceso de retorno (con respecto a un activo libre de riesgo) requerido para aceptar tener dicho activo dentro de su portafolio de activos riesgosos. A continuación se presenta la relación entre el retorno esperado de cualqueir activo y el de un portafolio frontera. 

##### Teorema 5

El retorno esperado de cualquier portafolio q (no necesariamente frontera) puede ser escrito como una combinación lineal del retorno esperado de un portafoliofrontera p y del de cero covarianza zc(p) con pesos $\beta_{qp}$ y $(1-\beta_{qp})$, es decir, 

$E[r_q] = \beta_{qp} E[r_p] + (1-\beta_{qp})E[r_{zc(p)}]$

A partir de esto se puede llegar a: 

$E[r_q-r_{zc(p)}] = \beta_{qp} E[r_p - r_{zc(p)}]$

Que nos dice que el exceo de retorno de cualquier portafolio q, no necesariamente frontera, con respecto al portafolio de cero covarianza de p es igual al exceso de retorno del portafolio frontera p, con respecto a su portafolio de cero covarianza, ponderado por una medida de riesgo o volatidiad, $\beta_qp$ (del portafolio q con respecto al portafolio referente p). 

Bajo supuesto débiles la expresión del teorema puede ser escrita como una regresión económetrica que nos da como resultado los siguientes coeficientes: 

$\beta_{qp} = \dfrac{Cov(r_q, r_p)}{\sigma^2(p)}$

$\beta_{qzc(p)} = \dfrac{Cov(r_q, zc(p))}{\sigma^2(zc(p))}$

Reorganizando nuestro cololario de forma que: 

$E[r_q] = E[r_{zc(p)}] + \beta_{qp} E[r_p - r_{zc(p)}]$

Y usando la definición de $\beta_{qp}$ dada por la regresión lineal podemos hacer algunas deducciones: si el portafolio p es eficiente entonces $E[r_p - r_{zc(p)}]$ es necesariamente positivo, así que el retorno esperado del portafolio q aumenta si está covariado con p y disminuye con la variabiliad de p. 


#### Análisis de la frontera con un activo libre de riesgo

A continuación dreivamos la frontera como un problema de elección de las ponderaciones óptimas de un portafolio p que consiste de n+1 activos (n riesgosos y un libre de riesgo). Definimos $\alpha_p'$ como un vector de n componentes que denota el peso de los activos riesgosos dentro del portafolio y rf al retorno del activo libre de riesgo. El problema a resolver es entonces 

$\displaystyle{\min_{\alpha_p} \dfrac{\alpha_p'v\alpha_p}{2} \space \space \space \text{s.a} \space \space \space \alpha_p'e+(1-\alpha_p')rf=E[rp]}$

El lagrangeano asociado al problema es entonces: 

$\displaystyle{\min_{\alpha_p, \lambda} L = \dfrac{\alpha_p'v\alpha_p}{2}  + \lambda(E[rp] - \alpha_p'e-(1-\alpha_p')rf)}$

Encontramos entonces la condiciones de primer orden del problema del inversor: 

$L_{\alpha_p} = v \alpha_p - \lambda e - \lambda Irf = 0$

$\implies \alpha_p = \lambda v^{-1} (e-Irf)$

$L_\lambda = E[rp] - \alpha_p'e-(1-\alpha_p'I)rf) = 0$

$\implies \alpha_p'(e-Irf)=E[rp]-rf$

Notar que es valido decir que: 

$\implies \alpha_p'v\alpha_p = \lambda \alpha_p'(e-Irf)$

Por lo cual 



$$
\begin{aligned}
\alpha_p^{\prime} v \overbrace{\alpha_p}^{=\lambda v^{-1}\left(e-I r_f\right)} &=\lambda^2\left(e-I r_f\right)^{\prime} v^{-1} v v^{-1}\left(e-I r_f\right), \\
&=\lambda^2\left(e-I r_f\right)^{\prime} v^{-1}\left(e-I r_f\right),
\end{aligned}
$$

Además, 

$$
\begin{aligned}
\alpha_p^{\prime} v \alpha_p &=\lambda \overbrace{\alpha_p^{\prime}\left(e-I r_f\right)}^{=E\left(r_p\right)-r_f}, \\
&=\lambda\left(E\left(r_p\right)-r_f\right) .
\end{aligned}
$$

Igualando las dos expresiones de la varianza derivadas anteriormente llegamos a una ecuación para $\lambda$

$$
\lambda=\underbrace{\frac{E\left(r_p\right)-r_f}{\left(e-I r_f\right)^{\prime} v^{-1}\left(e-I r_f\right)}}_{=H}=\frac{1}{H}\left(E\left(r_p\right)-r_f\right) .
$$

Finalmente, sustituyendo $\lambda$, obtenemos las ponderaciones óptimas

$\alpha_p = \dfrac{E[r_p]-r_f}{H} v^{-1}(e-Irf)$

##### La frontera cuando consideramos los activos riesgosos y el libre de riesgo: 

Para obtener una representación gráfica de la frontera procedemos a calcular la varianza del retorno del portafolio p: 

$\sigma^2\left(r_p\right) = \alpha_p^{\prime} V \alpha_p=\left(\frac{1}{H}\left(E\left(r_p\right)-r_f\right)\left(e-I r_f\right)^{\prime} V^{-1}\right) V\left(\frac{1}{H} V^{-1}\left(e-I r_f\right)\left(E\left(r_p\right)-r_f\right)\right) $

$= \frac{1}{H^2}\left(E\left(r_p\right)-r_f\right) \overbrace{\left(e-I r_f\right)^{\prime} V^{-1}\left(e-I r_f\right)\left(E\left(r_p\right)-r_f\right)}^H $

$= \frac{1}{H^2}\left(E\left(r_p\right)-r_f\right) H\left(E\left(r_p\right)-r_f\right)$

$\sigma^2\left(r_p\right)=\frac{1}{H}\left(E\left(r_p\right)-r_f\right)^2$

Por lo tanto, el desvío estándar es de la forma: 
$$
\sigma\left(r_p\right)=\left\{\begin{array}{ll}
\frac{1}{\sqrt{H}}\left(E\left(r_p\right)-r_f\right), & \text { si } E\left(r_p\right)-r_f \geq 0 \\
-\frac{1}{\sqrt{H}}\left(E\left(r_p\right)-r_f\right), & \text { si } E\left(r_p\right)-r_f \leq 0
\end{array}\right\}
$$

Como podemos notar, es una función partida donde sus dos partes son lineales que parten del punto (0,rf) con pendiente $\pm$ H. Graficamente luce de la siguiente forma:

#### Relación entre las fronteras con y sin el activo libre de riesgo 

Si juntamos ambos gráficos (con y sin activo libre de riesgo) tenemos lo siguiente: 

![frontera3.png](https://i.imgur.com/Jmjt5VF.png)

Al aumentar las posibilidades de inversión, la frontera con el activo libre de riesgo siempre va a estar por encima de la que no lo tiene. Además, claramente va a existir un punto de la frontera que también incluye al activo libre de riesgo, que va a ser común (tangente) a la frontera de los riesgosos. Este punto es el que le asigna peso cero al libre de riesgo y uno al portafolio de los riesgoso (e).

Vamos a ver que las decisiones con respecto a la cartera van a difereir dependiendo de la relación entre el retorno del portafolio de mínima varianza y la tada libre de riesgo (rf, A/C). Consideremos los siguientes casos: 

1. Cuando rf = A/C, el individuo invierte toda su riqueza inicial en el activo libre de riesgo. En este caso, el portafolio p es un portafolio de arbitraje (portafolio en el que no se invierte riqueza inicial).

2. Cuando rf < A/C, la frontera (del portafolio que incluye los activos riesgoss y el libre de riesgo) es tangente a la parte eficiente de la frontera del portafolio que solo incluye los activos riesgosos. Esta toca a la frontera de los activos en e. Notar que en este caso el portafolio e es eficiente y entonces todos los portafolios frontera (de la frontera que incluye el activo libre de riesgo) pueden ser generados como combinaciones lineales del activo libre de riesgo y el portafolio de tangencia, e. **No puede pasar desapercibido** que con esto volvemos a ganar separación monetaria de dos fondos. 

3. Cuando rf > A/C, el portafolio de tangencia e es ineficiente. El argumento es semejante al de 2. Pero en este caso, dado que e es ineficiente, el individuo debería shortear el portafolio e e ir long en el libre de riesgo. 

##### Relación entre el exceso de retorno de cualquier portafolio y el exceso de retorno de un portafolio frontera 

Usando las ponderaciones óptimas del portafolio $\alpha_p$ y un portafolio $\alpha_q$ cualquiera de activos riesgosos (solo riesgosos)podemos ver que la covarianza entre dichos portafolio es: 

$Cov(r_q, r_p) = \alpha_q'v\alpha_p = \dfrac{(E[r_q]-rf)(E[r_p]-rf)}{H}$

Despejando E[rq] lleagamos a que: 

$E[r_q]-rf=\beta_{qp}E[r_p - rf]$

Esta expresión nos dice que el retorno de cualquier portafolio (o activo) puede expresarse como una combinación lineal de la tasa libre de riesgo y un portafolio frontera. Vamos a ver un capitulo siguiente que esta relación va a ser la base del modelo de pricing más usado en finanzas.
